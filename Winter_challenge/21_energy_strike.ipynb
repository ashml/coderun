```python
import sys


def main():
    root = 3**0.5
    steps_x = steps_y = 8
    shifts = [(i / steps_x * 2, j / steps_y * root) for i in range(steps_x) for j in range(steps_y)]
    results = []
    for _ in range(int(input())):
        points = [tuple(map(float, input().split())) for _ in range(int(input()))]
        solution_found = False
        for shift_x, shift_y in shifts:
            possible = True
            current_centers = set()
            for x, y in points:
                found_node = False
                nx = x - shift_x
                ny = y - shift_y
                base_row = int(ny / root)
                best_x = best_y = 0
                for r in (base_row, base_row + 1):
                    row_y = r * root
                    base_col = int((nx - (r & 1)) * 0.5)
                    for c in (base_col, base_col + 1):
                        col_x = c * 2 + (r & 1)
                        dx = nx - col_x
                        dy = ny - row_y
                        if dx**2 + dy**2 <= 1:
                            best_cx = col_x + shift_x
                            best_cy = row_y + shift_y
                            found_node = True
                            break
                    if found_node: 
                        break
                
                if found_node:
                    current_centers.add((round(best_cx, 9), round(best_cy, 9)))
                else:
                    possible = False
                    break
            
            if possible:
                final_centers = list(current_centers)
                break

        out_lines = [f'{x} {y}' for x, y in final_centers]
        results.append(f'YES\n{len(final_centers)}\n' + '\n'.join(out_lines))
    print('\n'.join(results))


if __name__ == '__main__':
     main()
```
