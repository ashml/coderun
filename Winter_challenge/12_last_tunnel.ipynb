```python
import sys

MOD = 998244353

def mat_mul(A, B):
    size = 100
    res = [[0] * size for _ in range(size)]
    for i in range(size):
        for k in range(size):
            if A[i][k] == 0:
                continue
            for j in range(size):
                res[i][j] = (res[i][j] + A[i][k] * B[k][j]) % MOD
    return res

def vec_mul(v, mat):
    size = 100
    res = [0] * size
    for i in range(size):
        if v[i] == 0:
            continue
        for j in range(size):
            res[j] = (res[j] + v[i] * mat[i][j]) % MOD
    return res

def main():
    n, m = map(int, input().split())
    good_nums = list(map(int, input().split()))
    good = [False] * 28
    for a in good_nums:
        good[a] = True

    size = 100
    M = [[0] * size for _ in range(size)]
    for x in range(10):
        for y in range(10):
            i = x * 10 + y
            s = x + y
            row = M[i]
            for z in range(10):
                if good[s + z]:
                    j = y * 10 + z
                    row[j] = (row[j] + 1) % MOD

    v = [0] * size
    for d1 in range(1, 10):
        base = d1 * 10
        for d2 in range(10):
            v[base + d2] = 1

    p = n - 2
    while p > 0:
        if p & 1:
            v = vec_mul(v, M)
        M = mat_mul(M, M)
        p >>= 1

    ans = sum(v) % MOD
    print(ans)

if __name__ == '__main__':
    main()
```
