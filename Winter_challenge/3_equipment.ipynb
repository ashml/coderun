```python
import sys


def main():
    m, n = map(int, input().split())
    w = list(map(int, input().split()))
    mod = 10**9 + 7

    left, right = 0, max(w)

    while left < right:
        mid = (left + right) // 2
        if sum(max(val - mid, 0) for val in w) >= m:
            left = mid + 1
        else:
            right = mid

    leftover = sum(max(val - left + 1, 0) for val in w) - m
    cost = sum(min(val, left - 1)**2 for val in w) % mod
    add = (2 * (left - 1) + 1) * leftover
    answer = (cost + add) % mod
    print(answer)


if __name__ == '__main__':
    main()
```
