```python
import sys
from array import array
from collections import defaultdict


def new_node(left, right, cnt, val):
    left.append(-1)
    right.append(-1)
    cnt.append(0)
    val.append(0)
    return len(left) - 1

def update(x, add, left, right, cnt, val):
    node = 0
    path = []

    for bit in range(29, -1, -1):
        path.append(node)
        b = (x >> bit) & 1
        if b == 0:
            if left[node] == -1:
                left[node] = new_node(left, right, cnt, val)
            node = left[node]
        else:
            if right[node] == -1:
                right[node] = new_node(left, right, cnt, val)
            node = right[node]

    cnt[node] = 1 if add else 0
    val[node] = 1 if add else 0

    for level in range(30):
        cur = path[29 - level]
        size = 1 << level

        l = left[cur]
        r = right[cur]

        cl = cnt[l] if l != -1 else 0
        cr = cnt[r] if r != -1 else 0

        cnt[cur] = cl + cr

        vl = val[l] if l != -1 else 0
        vr = val[r] if r != -1 else 0

        full_l = cl == size
        full_r = cr == size

        if full_l and full_r:
            val[cur] = size << 1
        elif full_l:
            val[cur] = size + vr
        elif full_r:
            val[cur] = size + vl
        else:
            val[cur] = max(vl, vr)

def main():
    t = int(input())
    output = []
    for _ in range(t):
        n, q = map(int, input().split())
        a = list(map(int, input().split()))

        left = array('i', [-1])
        right = array('i', [-1])
        cnt = array('i', [0])
        val = array('i', [0])

        freq = defaultdict(int)

        for x in a:
            if freq[x] == 0:
                update(x, True, left, right, cnt, val)
            freq[x] += 1

        output.append(str(val[0]))

        for _ in range(q):
            i, x = map(int, input().split())
            i -= 1

            old = a[i]
            if old != x:
                a[i] = x

                freq[old] -= 1
                if freq[old] == 0:
                    update(old, False, left, right, cnt, val)

                if freq[x] == 0:
                    update(x, True, left, right, cnt, val)
                freq[x] += 1

            output.append(str(val[0]))

    print('\n'.join(output))


if __name__ == "__main__":
    main()
```
