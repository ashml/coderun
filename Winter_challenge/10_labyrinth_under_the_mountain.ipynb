```python
import sys
from collections import defaultdict


def main():
    input = sys.stdin.readline
    t = int(input())
    for _ in range(t):
        n, m = map(int, input().split())
        a = list(map(int, input().split()))
        b = list(map(int, input().split()))
        c = list(map(int, input().split()))
        
        edges = []
        for i in range(m):
            if a[i] != b[i]:
                edges.append((c[i], a[i], b[i]))
        
        edges.sort()
        
        parent = list(range(n + 1))
        rank = [0] * (n + 1)
        size = [1] * (n + 1)
        
        def find(x):
            if parent[x] != x:
                parent[x] = find(parent[x])
            return parent[x]
        
        def union(x, y):
            px, py = find(x), find(y)
            if px == py:
                return 0
            if rank[px] < rank[py]:
                px, py = py, px
            parent[py] = px
            size[px] += size[py]
            if rank[px] == rank[py]:
                rank[px] += 1
            return size[px]
        
        max_component = defaultdict(int)
        max_component[0] = 1
        
        for weight, u, v in edges:
            new_size = union(u, v)
            if new_size > 0:
                if max_component[weight] < new_size:
                    max_component[weight] = new_size
        
        result = [-1] * (n + 1)
        result[1] = 0
        
        current_max = 1
        sorted_weights = sorted(max_component.keys())
        
        for w in sorted_weights:
            if max_component[w] > current_max:
                for k in range(current_max + 1, max_component[w] + 1):
                    if result[k] == -1:
                        result[k] = w
                current_max = max_component[w]
        
        print(' '.join(map(str, result[1:n+1])))


if __name__ == '__main__':
    main()
```
