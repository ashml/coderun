```python
import sys

def main():
    MOD = 10**9 + 7

    def sum1(n): 
        return n * (n + 1) // 2
        
    def sum2(n): 
        return n * (n + 1) * (2 * n + 1) // 6
    
    a, q, L, R = map(int, input().split())
    N = R - L + 1

    if N <= 1 or a == 0 or q == 1:
        print(0)
        return

    if q == 0:
        if not (L <= 0 <= R):
            print(0)
        else:
            print((2 * (N - 1) * N**2) % MOD)
        return

    if q == -1:
        cnt_even = N // 2 + (1 if (N % 2 != 0 and L % 2 == 0) else 0)
        cnt_odd = N - cnt_even
        ans = 2 * cnt_even * cnt_odd * N**2
        print(ans % MOD)
        return
  
    M = N - 1
    term1 = N**2 * (N - 1)
    c0 = 2 * N * (N + 1)
    c1 = -2 * (2 * N + 1)
    total_0 = 0 
    total_1 = 0
    total_2 = 0

    l = 1
    while l <= M:
        qval = M // l
        r = M // qval

        total_0 += (r - l + 1) * qval
        total_1 += (sum1(r) - sum1(l - 1)) * sum1(qval)
        total_2 += (sum2(r) - sum2(l - 1)) * sum2(qval)
        l = r + 1

    term2 = 2 * (total_2 - (2 * N + 1) * total_1 + N * (N + 1) * total_0)

    if q == -2:
        K = M // 2
        all_x_sum = (c0 * M + c1 * sum1(M) + 2 * sum2(M)) % MOD
        sum_even_x = (c0 * K + c1 * 2 * sum1(K) + 2 * 4 * sum2(K)) % MOD
        sum_odd_x = (all_x_sum - sum_even_x + MOD) % MOD
        p_at_1 = (c0 + c1 + 2) % MOD
        correct_val = 2 * (N - 2) * (N + 1) % MOD
        delta = (sum_odd_x - p_at_1 + correct_val + MOD) % MOD
        term2 = (term2 + delta) % MOD

    print((term1 + term2) % MOD)

if __name__ == '__main__':
    main()
```
