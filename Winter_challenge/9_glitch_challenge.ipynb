```python
import sys
from collections import defaultdict

def count_dividers(n, size, spf, exp):
    MOD = 10**9 + 7
    dividers = 1
    for p in range(2, size):
        if spf[p] == p:
            v = 0
            pp = p
            while pp <= n:
                v += n // pp
                pp *= p
            v -= exp[p]
            if v < 0:
                return 0
            dividers = (dividers * (v + 1)) % MOD
    return dividers

def main():
    n, k = map(int, input().split())
    a = list(map(int, input().split()))
    size = min(n, 10**6) + 1
    
    spf = list(range(size))
    exp = defaultdict(int)
    
    for i in range(2, int(size**0.5) + 1):
        if spf[i] == i:
            step = i
            start = i**2
            for j in range(start, size, step):
                if spf[j] == j:
                    spf[j] = i

    for val in a:
        while val > 1:
            cnt = 0
            p = spf[val]
            while val % p == 0:
                val //= p
                cnt += 1
            exp[p] += cnt

    print(count_dividers(n, size, spf, exp))


if __name__ == '__main__':
    main()
```
