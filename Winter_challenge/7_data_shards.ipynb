```python
import sys


mod = 998244353

def main():
    queries = [tuple(map(int, input().split())) for _ in range(int(input()))]
    max_s = max(s for _, s in queries)

    fact = [1] * (max_s + 1)
    for i in range(1, max_s + 1):
        fact[i] = fact[i - 1] * i % mod

    inv_fact = [1] * (max_s + 1)
    inv_fact[max_s] = pow(fact[max_s], -1, mod)
    for i in range(max_s, 0, -1):
        inv_fact[i - 1] = inv_fact[i] * i % mod

    for n, s in queries:
        answer = 0
        if s >= n:
            answer = (n + 1) * fact[s] % mod * inv_fact[s - n] % mod
        print(answer)


if __name__ == '__main__':
    main()
```
