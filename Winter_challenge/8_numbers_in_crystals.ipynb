```python
import sys


def main():
    def dfs(start, cur_prod, depth):
        for x in range(start, max_r // cur_prod + 1):
            p = cur_prod * x
            if good_k[p] < depth + 1:
                good_k[p] = depth + 1
            if depth + 1 < max_k:
                dfs(x + 1, p, depth + 1)
                
    queries = []
    max_r = 0
    for _ in range(int(input())):
        k, l, r = map(int, input().split())
        queries.append((k, l, r))
        max_r = max(max_r, r)

    max_k = 0
    prod = 1
    cur = 2
    while prod * cur <= max_r:
        prod *= cur
        max_k += 1
        cur += 1

    good_k = [0] * (max_r + 1)
    dfs(2, 1, 0)
    prefix = [[0] * (max_r + 1) for _ in range(max_k + 1)]

    for k in range(1, max_k + 1):
        s = 0
        for i in range(1, max_r + 1):
            if good_k[i] >= k:
                s += 1
            prefix[k][i] = s

    answer = []
    for k, l, r in queries:
        if k > max_k:
            answer.append('0')
        else:
            answer.append(str(prefix[k][r] - prefix[k][l - 1]))
    print('\n'.join(answer))


if __name__ == '__main__':
    main()
```
