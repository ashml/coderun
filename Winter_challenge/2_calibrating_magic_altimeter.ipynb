```python
import sys


def main():
    values = [int(input()) for _ in range(10)]
    best_sum = 0
    best_diff = 100

    for mask in range(1 << 10):
        current_sum = 0
        for i in range(10):
            if mask & (1 << i):
                current_sum += values[i]
        diff = abs(current_sum - 100)
        if diff < best_diff or (diff == best_diff and current_sum > best_sum):
            best_diff = diff
            best_sum = current_sum

    print(best_sum)


if __name__ == '__main__':
    main()
```

```python
import sys


def dp(numbers):
    target = 100
    max_possible = 1000
    dp = [False] * (max_possible + 1)
    dp[0] = True

    for value in numbers:
        for s in range(max_possible, value - 1, -1):
            if dp[s - value]:
                dp[s] = True

    for i in range(0, max_possible + 1):
        higher = target + i
        lower = target - i
        if higher <= max_possible and dp[higher]:
            return higher
        if lower >= 0 and dp[lower]:
            return lower


def main():
    numbers = [int(input()) for _ in range(10)]
    max_sum = sum(numbers)
    if max_sum <= 100:
        print(max_sum)
    else:
        print(dp(numbers))
    

if __name__ == '__main__':
    main()
```
