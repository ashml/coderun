```python
import sys


def main():
    R, B = map(int, input().split())
    h = 1
    while h**2 <= B:
        if B % h == 0:
            W = (B // h) + 2
            H = h + 2
            if 2 * W + 2 * H - 4 == R:
                print(max(W, H), min(W, H))
                break
        h += 1

if __name__ == '__main__':
    main()
```
