```python
import sys


def main():
    MOD = 10**9 + 7
    m = int(input())
    a = list(map(int, input().split()))
    n = sum(a)
    k = 0
    for el in a:
        k = max(k, el)

    inv = [1] * n
    targets = {k - 1, n - k, n - 1}
    H_vals = {0: 0}
    curr_H = 1
    if 1 in targets:
        H_vals[1] = curr_H
        
    for i in range(2, n):
        inv[i] = (-(MOD // i) * inv[MOD % i]) % MOD
        curr_H = (curr_H + inv[i]) % MOD
        if i in targets:
            H_vals[i] = curr_H

    term = (H_vals[n - 1] - H_vals[k - 1] + H_vals[n - k]) % MOD
    ans = (n - 1) * ((MOD + 1) // 2) % MOD * term % MOD
    print(ans)


if __name__ == '__main__':
    main()
```
