```python
def main():
    n = int(input())
    p = list(map(int, input().split()))
    
    q = [0] * n
    for k in range(n // 2):
        pos_1 = 2 * k
        pos_2 = 2 * k + 1
        val_1 = 2 * k + 1
        val_2 = 2 * k + 2
        
        if p[pos_1] != val_1 and p[pos_2] != val_2:
            q[pos_1] = val_1
            q[pos_2] = val_2
        else:
            q[pos_1] = val_2
            q[pos_2] = val_1
    
    if n % 2 and p[n - 1] != n:
        q[n - 1] = n
    elif n % 2:
        q[n - 1] = q[n - 2]
        q[n - 2] = n
    
    print(' '.join(map(str, q)))

if __name__ == '__main__':
    main()
```
