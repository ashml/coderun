```python
import sys


def main():
    n = int(input())
    mod = 10**9 + 7
    ans = 0
    
    inv = [0] * (n + 1)
    inv[1] = 1
    for i in range(2, n + 1):
        inv[i] = (mod - mod // i) * inv[mod % i] % mod
        
    d = [0] * (n + 2)
    d[n] = 1
    for k in range(n - 1, 0, -1):
        d[k] = (n + (n - k) * d[k + 1]) % mod
        d[k] = d[k] * inv[k] % mod
        
    for k in range(1, n + 1):
        ans = (ans + d[k]) % mod
        
    print(ans)


if __name__ == '__main__':
    main()
```
