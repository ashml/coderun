```python
import sys


def main():
    def find(a):
        while a != parent[a]:
            parent[a] = parent[parent[a]]
            a = parent[a]
        return a
    
    def union(a, b):
        a = find(a)
        b = find(b)
        if a == b:
            return False
        if rank[a] < rank[b]:
            parent[a] = b
        else:
            parent[b] = a
            if rank[a] == rank[b]:
                rank[a] += 1
        return True
    
    n = int(input())
    points = [(i, *map(int, input().split())) for i in range(n)]
    edges = []
    for dim in range(1, 4):
        points.sort(key=lambda p: p[dim])
        for i in range(n - 1):
            i1, x1, y1, z1 = points[i]
            i2, x2, y2, z2 = points[i + 1]
            cost = min(abs(x1 - x2), abs(y1 - y2), abs(z1 - z2))
            edges.append((cost, i1, i2))

    edges.sort()
    parent = list(range(n))
    rank = [0] * n
    answer = 0
    cnt = 0
    for cost, a, b in edges:
        if union(a, b):
            answer += cost
            cnt += 1
            if cnt == n - 1:
                break
    print(answer)


if __name__ == '__main__':
    main()
```
