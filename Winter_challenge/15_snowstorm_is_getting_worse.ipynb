```python
import sys


def main():
    mod = 998244353
    for _ in range(int(input())):
        ans = 0
        s = input()
        t = input()
        n, m = len(s), len(t)

        s_chars = [ord(c) for c in s]
        t_chars = [ord(c) for c in t]
        
        part = [1] * (n + 1)
        lcp = [[0] * (m + 1) for _ in range(n + 1)]

        for i in range(1, n):
            part[i + 1] = (part[i] << 1) % mod

        for i in range(n - 1, -1, -1):
            for j in range(m - 1, -1, -1):
                lcp[i][j] = lcp[i + 1][j + 1] + 1 if s_chars[i] == t_chars[j] else 0

        for start in range(m):
            rem = m - start
            eq = [0] * (n + 1)
            eq[n] = 1

            for pos in range(n, 0, -1):
                len_pre = n - pos
                t_pos = start + len_pre
                if eq[pos] == 0 or len_pre >= rem:
                    continue

                for k in range(pos - 1, -1, -1):
                    j = lcp[k][t_pos]
                    if j >= pos - k:
                        if pos - k < rem - len_pre:
                            eq[k] += eq[pos]
                    elif j < rem - len_pre and s_chars[k + j] < t_chars[t_pos + j]:
                        ans = (ans + eq[pos] * part[k] * (rem - len_pre - j)) % mod
            ans = (ans + eq[0] * max(0, rem - n)) % mod
        print(ans % mod)


if __name__ == '__main__':
    main()
```
