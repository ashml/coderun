```python
import sys

mod = 998244353

def main():
    t = int(input())

    max_n = 3000
    stirling = [[0] * (max_n + 1) for _ in range(max_n + 1)]
    stirling[0][0] = 1
    for i in range(1, max_n + 1):
        for j in range(1, i + 1):
            stirling[i][j] = (stirling[i - 1][j - 1] + (i - 1) * stirling[i - 1][j]) % mod

    results = []
    for _ in range(t):
        n, q, l, r = map(int, input().split())
        b = [int(x) - 1 for x in input().split()]
        c = [int(x) - 1 for x in input().split()]

        next_station = [-1] * n
        indegree = [0] * n
        valid = True

        for i in range(q):
            u, v = b[i], c[i]
            if next_station[u] != -1 and next_station[u] != v:
                valid = False
                break
            next_station[u] = v
            indegree[v] += 1
            if indegree[v] > 1:
                valid = False
                break

        if not valid:
            results.append('0')
            continue

        visited = [0] * n
        fixed_cycles = 0

        for i in range(n):
            if next_station[i] != -1 and visited[i] == 0:
                current = i
                stack = {}
                while current != -1 and visited[current] == 0:
                    stack[current] = len(stack)
                    visited[current] = 1
                    current = next_station[current]
                if current in stack:
                    fixed_cycles += 1

        chains = sum(1 for i in range(n) if next_station[i] != -1 and indegree[i] == 0)
        isolated = sum(1 for i in range(n) if next_station[i] == -1 and indegree[i] == 0)

        free = chains + isolated
        ans = 0
        for x in range(free + 1):
            total_cycles = fixed_cycles + x
            if l <= total_cycles <= r:
                ans = (ans + stirling[free][x]) % mod

        results.append(str(ans))

    print('\n'.join(results))

if __name__ == '__main__':
    main()
```
