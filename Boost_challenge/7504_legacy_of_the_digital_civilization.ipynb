```python
def solution(n: int, m: int) -> int:
    def fits(width):
        total = rank = 0
        comb = 1
        while comb < width and rank <= n:
            total += comb
            rank += 1
            comb = comb * (n - rank + 1) // rank
        total = 2 * total + max(0, n + 1 - 2 * rank) * width
        return total >= m

    low = (m + n) // (n + 1)
    high = m
    while low < high:
        mid = (low + high) // 2
        if fits(mid):
            high = mid
        else:
            low = mid + 1
    return low
```

```python
print(solution(1, 2))  # 1
print(solution(5, 6))  # 1
print(solution(3, 5))  # 2
```
