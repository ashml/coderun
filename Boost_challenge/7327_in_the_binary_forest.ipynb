```python
def solve(n: int, a: list[int], m: int, b: list[int]) -> int:
    max_len = 0
    zero_pos_a = [i for i, x in enumerate(a) if x == 0]
    zero_pos_b = [i for i, x in enumerate(b) if x == 0]
    max_zeros = min(len(zero_pos_a), len(zero_pos_b))
    
    ones_after_a = [0] * (n + 1)
    ones_after_b = [0] * (m + 1)
    
    for i in range(n - 1, -1, -1):
        ones_after_a[i] = ones_after_a[i + 1] + (a[i] == 1)
        
    for j in range(m - 1, -1, -1):
        ones_after_b[j] = ones_after_b[j + 1] + (b[j] == 1)

    for i in range(max_zeros + 1):
        last_a = zero_pos_a[i - 1] if i > 0 else -1
        last_b = zero_pos_b[i - 1] if i > 0 else -1
        ones = min(ones_after_a[last_a + 1], ones_after_b[last_b + 1])
        max_len = max(max_len, i + ones)
    return max_len
```

```python
print(solve(7, [0, 0, 0, 1, 0, 1, 1], 6, [0, 0, 1, 1, 0, 1]))  # 5
```
