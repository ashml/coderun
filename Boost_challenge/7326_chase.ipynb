```python
from collections import deque

def solve(n: int, k: int, a: list[int]) -> list[int]:
    queue = deque([0])
    previous = [-1] * (n + 1)
    jumps = [0] * (n + 1)

    visited = [False] * (n + 1)
    visited[0] = True

    reachable = [False] * (n + 1)
    reachable[n] = True
    for x in a:
        reachable[x] = True

    while queue:
        position = queue.popleft()
        if position == n:
            break
        for step in (1, 2):
            next_pos = position + step
            if next_pos <= n and reachable[next_pos] and not visited[next_pos]:
                visited[next_pos] = True
                previous[next_pos] = position
                jumps[next_pos] = step
                queue.append(next_pos)
    if not visited[n]:
        return [-1]
        
    path = []
    current = n
    while current != 0:
        path.append(jumps[current])
        current = previous[current]
    path.reverse()
    return [len(path)] + path
```
