```python
import sys
from io import StringIO

class FastInput:
    def __init__(self):
        self.stdin = sys.stdin

    def read_line(self):
        return sys.stdin.readline().strip()

    def read_tokens(self):
        return self.read_line().split()

    def read_int(self):
        return int(self.read_line())

    def read_ints(self):
        return map(int, self.read_tokens())


# ввод/вывод
# не изменяйте сигнатуру метода
def solution():
    reader = FastInput()
    t = reader.read_int()
    output = []

    for _ in range(t):
        n = reader.read_int()
        permutation = [x - 1 for x in reader.read_ints()]
        costs = list(reader.read_ints())

        visited = [False] * n
        cycle_count = 0
        total_min_cost = 0

        for start in range(n):
            if not visited[start]:
                cycle_count += 1
                current = start
                cycle_min_cost = float('inf')
                while not visited[current]:
                    visited[current] = True
                    cycle_min_cost = min(cycle_min_cost, costs[current])
                    current = permutation[current]
                total_min_cost += cycle_min_cost

        output.append('0' if cycle_count <= 1 else str(total_min_cost))

    print('\n'.join(output))
```

```python
example_input = """\
5
1
1
2
4
2 3 4 1
4 2 5 5
3
1 2 3
5 4 7
3
1 2 3
0 0 0
6
2 3 1 5 4 6
1 1 1 1 1 2
"""

stdin_backup, stdout_backup = sys.stdin, sys.stdout
sys.stdin = StringIO(example_input)
sys.stdout = StringIO()

solution()
output = sys.stdout.getvalue()
sys.stdin, sys.stdout = stdin_backup, stdout_backup

print(output)
```
