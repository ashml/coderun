```python
def solution(n: int, m: int, p: list[int]) -> list[int]:
    known = [(0, 0)]
    answer = [0] * n
    for i, val in enumerate(p):
        if val != -1:
            known.append((i + 1, val))
            
    for i in range(len(known)):
        left_i, left_val = known[i]
        if i == len(known) - 1:
            for j in range(left_i, n):
                answer[j] = m
            break

        rigth_i, rigth_val = known[i + 1]
        dist = rigth_i - left_i
        coins_delta = rigth_val - left_val
        
        if coins_delta < dist * m:
            return [-1]
            
        for j in range(left_i + 1, rigth_i):
            answer[j - 1] = m
        answer[rigth_i - 1] = coins_delta - m * (dist - 1)
    return answer
```

```python
print(solution(3, 1, [1, 2, 3]))    # [1, 1, 1]
print(solution(2, 0, [-1, -1]))     # [0, 0]
print(solution(4, 2, [2, -1, -1, 5]))
```
