```python
import sys


def calculate_answer(n: int, a: list[int]):
    adj = [x - 1 if x != -1 else -1 for x in a]
    in_degree = [0] * n
    for dest in adj:
        if dest != -1:
            in_degree[dest] += 1
            
    subtree_size = [1] * n
    queue = [i for i, d in enumerate(in_degree) if d == 0]
    while queue:
        u = queue.pop()
        v = adj[u]
        if v != -1:
            subtree_size[v] += subtree_size[u]
            in_degree[v] -= 1
            if in_degree[v] == 0:
                queue.append(v)
                
    ans = list(subtree_size)
    for i in range(n):
        if in_degree[i] > 0:
            cycle_nodes = []
            curr = i
            while in_degree[curr] > 0:
                in_degree[curr] = 0
                cycle_nodes.append(curr)
                curr = adj[curr]
            
            cycle_total = sum(subtree_size[node] for node in cycle_nodes)
            for node in cycle_nodes:
                ans[node] = cycle_total
                
    return ans

# ввод/вывод внизу

class FastInput:
    def __init__(self):
        self.stdin = sys.stdin

    def read_line(self):
        return sys.stdin.readline().strip()

    def read_tokens(self):
        return self.read_line().split()

    def read_int(self):
        return int(self.read_line())

    def read_ints(self):
        return map(int, self.read_tokens())


# ввод/вывод
# не изменяйте сигнатуру метода
def solution():
    input = FastInput()

    t = input.read_int()

    answers = [None] * t

    for test in range(t):

        n = input.read_int()
        a = list(input.read_ints())

        answers[test] = calculate_answer(n, a)

    print('\n'.join(' '.join(map(str, ans)) for ans in answers))
```
