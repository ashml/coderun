```python
import sys


def compute_mex(a):
    s = set()
    mex = 0
    for el in a:
        if el <= len(a) + 1:
            s.add(el)
    while mex in s:
        mex += 1
    return mex


def main():
    t = int(input())
    out_lines = []
    for _ in range(t):
        n = int(input())
        a = list(map(int, input().split()))
        mex_total = compute_mex(a)
        max_val = max(a)
        min_val = min(a)

        l, r = 0, max_val + 2
        while l < r:
            m = (l + r) // 2
            left = right = -1
            for i, x in enumerate(a):
                if x > m:
                    if left == -1:
                        left = i
                    right = i
            if left == -1:
                valid = (min_val != 0 or m >= 1)
            else:
                segment = a[left:right + 1]
                valid = compute_mex(segment) <= m
            if valid:
                r = m
            else:
                l = m + 1

        min_max_val = l
        count_max = a.count(max_val)
        max_max_val = max(mex_total, max_val if count_max >= 2 or n >= 2 else -1)
        min_min_val = 1 if n == 1 and a[0] == 0 else 0
        max_min_val = mex_total
        out_lines.append(f"{min_max_val} {max_max_val} {min_min_val} {max_min_val}")
        
    for line in out_lines:
        print(line)


if __name__ == '__main__':
    main()
```
