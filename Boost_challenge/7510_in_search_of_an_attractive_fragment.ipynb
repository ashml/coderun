```python
import sys

def calculate_answer(n: int, s: str) -> int:
    mod = 998244353
    inv_2 = pow(2, mod - 2, mod)
    segments = []
    for start in range(n):
        for end in range(start + 1, n, 2):
            segments.append((start, end))
    
    total_segments = len(segments)
    inv_total_segments = pow(total_segments, mod - 2, mod)
    
    memo = {}

    def solve_state(current_mask):
        if current_mask in memo:
            return memo[current_mask]
        
        stay_probability = 0
        transition_sum = 0
        for left, right in segments:
            current_prob_bad = 1
            is_found = False
            l_ptr, r_ptr = left, right
            while l_ptr < r_ptr:
                bit_l = (current_mask >> l_ptr) & 1
                bit_r = (current_mask >> r_ptr) & 1
                if bit_l or bit_r:
                    current_prob_bad = (current_prob_bad * inv_2) % mod
                elif s[l_ptr] == s[r_ptr]:
                    is_found = True
                    break
                
                l_ptr += 1
                r_ptr -= 1
            
            if is_found:
                continue
            
            sub_length = right - left + 1
            impact_mask = ((1 << sub_length) - 1) << left
            next_mask = current_mask | impact_mask
            
            if next_mask == current_mask:
                stay_probability = (stay_probability + current_prob_bad) % mod
            else:
                res = solve_state(next_mask)
                transition_sum = (transition_sum + current_prob_bad * res) % mod

        denominator = (total_segments - stay_probability + mod) % mod
        expected_value = ((transition_sum + total_segments) * pow(denominator, mod - 2, mod)) % mod
        memo[current_mask] = expected_value
        return expected_value

    return solve_state(0)

class FastInput:
    def __init__(self):
        self.stdin = sys.stdin

    def read_line(self):
        return sys.stdin.readline().strip()

    def read_tokens(self):
        return self.read_line().split()

    def read_int(self):
        return int(self.read_line())

    def read_ints(self):
        return map(int, self.read_tokens())

def solution():
    input = FastInput()

    s = input.read_line()
    n = len(s)

    answer = calculate_answer(n, s)
    print(answer)

if __name__ == "__main__":
    solution()
```
