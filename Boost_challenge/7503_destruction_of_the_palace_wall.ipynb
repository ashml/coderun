```python
def solve(n: int, a: list[int]) -> int:
    max_stack = []
    min_stack = []
    ans = 0
    for i, x in enumerate(a + [0]):
        while max_stack and (i == n or a[max_stack[-1]] < x):
            j = max_stack.pop()
            k = max_stack[-1] if max_stack else -1
            ans += a[j] * (j - k) * (i - j)
        max_stack.append(i)
        while min_stack and (i == n or a[min_stack[-1]] > x):
            j = min_stack.pop()
            k = min_stack[-1] if min_stack else -1
            ans -= a[j] * (j - k) * (i - j)
        min_stack.append(i)
    return ans
```

```python
print(solve(1, [1]))           # 0
print(solve(2, [1, 11]))       # 10
print(solve(3, [1, 3, 5]))     # 8
```
