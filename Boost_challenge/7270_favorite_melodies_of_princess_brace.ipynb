```python
def solution(n: int, q: int, a, queries) -> int:
    cnt = [0] * (n + 2)
    for start, end in queries:
        cnt[start] += 1
        cnt[end + 1] -= 1

    for i in range(1, n + 1):
        cnt[i] += cnt[i - 1]

    return sum(joi * freq for joi, freq in zip(sorted(a), sorted(cnt[1:n + 1])))
```

```python
print(solution(3, 4, [7, 3, 1], [[1, 3], [2, 3], [3, 3], [2, 2]]))
```
