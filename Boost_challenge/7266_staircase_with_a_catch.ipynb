```python
def solution(n, a):
    a[0] = min(a[0], -a[0])
    for i in range(1, n):
        if a[i] >= a[i-1] and -a[i] >= a[i-1]:
            a[i] = min(a[i], -a[i])
        elif -a[i] >= a[i-1]:
            a[i] = -a[i]
        elif a[i] < a[i-1] and -a[i] < a[i-1]:
            return [0]
    return [1] + a
```

```python
print(solution(5, [1, -1, -2, 3, 6]))   # [1, -1, -1, 2, 3, 6]
print(solution(3, [1, 1, 0]))           # [1, -1, -1, 0]
print(solution(5, [5, 2, 1, 2, 1]))     # [0]
```
