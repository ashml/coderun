```python
class Answer:
    _slots_ = [
        'sum', 
        'field',
    ]

    def __init__(self, sum, field):
        self.sum = sum
        self.field = field

def solution(n):
    total_sum = (n - 1) * (3 * n - 2)
    field = [
        "x" * n if i % 2 == 0 else "-" * n
        for i in range(n)
    ]

    return Answer(
        sum=total_sum,
        field=field,
    )
```
