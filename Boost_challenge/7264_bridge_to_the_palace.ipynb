```python
def solution(n, a):
    s = sorted(a)
    left = 0
    max_in_window = 0
    for right in range(n):
        while s[right] - s[left] >= n:
            left += 1
        max_in_window = max(max_in_window, right - left + 1)
    return n - max_in_window
```

```python
solution(5, [2, 0, -3, 3, 6])
```
