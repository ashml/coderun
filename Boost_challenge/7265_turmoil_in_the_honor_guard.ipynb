```python
def solution(n, m, swaps):
    pos = [i for i in range(2*n+1)]
    curr_methods = n
    result = []
    for k in range(m):
        a, b = swaps[2*k:2*k+2]
        curr_methods -= (a <= n and pos[a] <= n) + (b <= n and pos[b] <= n)
        curr_methods += (a <= n and pos[b] <= n) + (b <= n and pos[a] <= n)
        pos[a], pos[b] = pos[b], pos[a]
        result.append(curr_methods)
    return result
```

```python
n = 3 
m = 3
swaps = [3, 4, 4, 5, 5, 3]
```

```python
solution(n, m, swaps)
```
