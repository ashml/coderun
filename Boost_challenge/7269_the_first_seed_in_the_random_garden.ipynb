```python
import sys
from functools import lru_cache
sys.setrecursionlimit(10**7)


@lru_cache
def solution(n: int) -> int:
    MOD = 10**9 - 7538
    if n == 0:
        return 1
    term1 = pow(solution(n // 2), solution(n // 3), MOD)
    term2 = 5 * solution(n // 4)
    return (term1 + term2 + n) % MOD
```

```python
print(solution(1), '->', 7)
print(solution(5), '->', 105413544)
print(solution(100), '->', 464980765)
```
