```python
from collections import defaultdict

def solution(n: int, words: list[str]) -> list[list[int]]:
    indexed_words = [(word, i) for i, word in enumerate(words)]
    indexed_words.sort(key=lambda x: len(x[0]))
    used = [False] * (2 * n)
    prefix_map = defaultdict(list)
    for word, idx in indexed_words:
        prefix_map[word].append(idx)
    result = []
    for word, idx in reversed(indexed_words):
        if used[idx]:
            continue
        for k in range(len(word), 0, -1):
            prefix = word[:k]
            if prefix not in prefix_map:
                continue
            while prefix_map[prefix]:
                s_idx = prefix_map[prefix].pop()
                if s_idx != idx and not used[s_idx]:
                    result.append([s_idx + 1, idx + 1])
                    used[s_idx] = True
                    used[idx] = True
                    break
            if used[idx]:
                break
    return result
```

```python
n = 3
words = [
    "sunshine",
    "sun",
    "doghouse",
    "dog",
    "caterpillar",
    "cat"
]
res = solution(n, words)
for a, b in res:
    print(a, b)
```
