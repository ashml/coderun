```python
def get_primes(n):
    if n < 2:
        return []
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False
    for i in range(2, int(n ** 0.5) + 1):
        if is_prime[i]:
            for j in range(i * i, n + 1, i):
                is_prime[j] = False
    return [i for i, prime in enumerate(is_prime) if prime and i != 2]

def solution(n: int) -> int:
    mod_1 = 0
    mod_3 = 0
    for el in get_primes(n):
        if el % 4 == 1:
            mod_1 += 1
        else:
            mod_3 += 1
    return mod_1 * mod_3
```

```python
print(solution(3), '->', 0)
print(solution(5), '->', 1)
print(solution(7), '->', 2)
print(solution(20), '->', 12)
```
