```python
def rain(n, t, a, b):
    result = [0] * (t + 1)
    total = sum(a)
    result[0] = total

    delta_speed = [0] * (t + 2)
    delta_rem   = [0] * (t + 2)

    for i in range(n):
        if b[i] == 0:
            continue
        delta_speed[1] += b[i]
        lifetime = a[i] // b[i]
        end_time = lifetime + 1
        if end_time <= t:
            delta_speed[end_time] -= b[i]
            delta_rem[end_time] += a[i] - b[i] * lifetime

    cur_speed = 0
    for minute in range(1, t + 1):
        cur_speed += delta_speed[minute]
        total -= cur_speed
        total -= delta_rem[minute]
        result[minute] = max(total, 0)

    return result
```

```python
n = 6
t = 6
a = [12, 10, 3, 5, 7, 9]
b = [2, 3, 1, 4, 2, 1]
expected = [46, 33, 23, 14, 9, 6, 3]
result = rain(n, t, a, b)
print(result, '->', expected)
```
