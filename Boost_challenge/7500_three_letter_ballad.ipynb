```python
from collections import Counter

def solve(ballad: str, n: int) -> int:
    ballad = ballad.replace(' ', '')
    answer = 0
    n_comb = 0
    left = Counter()
    right = Counter(ballad)
    for el in ballad:
        right[el] -= 1
        n_comb -= left[el]
        answer += n_comb
        n_comb += right[el]
        left[el] += 1
    return answer
```

```python
print(solve("treasure", 8))                          # 8
print(solve("you will never find the treasure", 32))  # 146
```
