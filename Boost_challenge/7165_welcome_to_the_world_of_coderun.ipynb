```python
def solution(n, m):
    def is_possible(mid):
        total = mid**2
        if m + n < total:
            return False
        white = total // 2
        black = (total + 1) // 2
        return (white <= n and black <= m) or (white <= m and black <= n)

    answer = 0
    left = 0
    right = 10**9
    while left <= right:
        mid = (left + right) // 2
        if is_possible(mid):
            answer = mid
            left = mid + 1
        else:
            right = mid - 1
    return answer
```

```python
print(solution(9, 8))
print(solution(12, 15))
```
