```python
def solve(a: int, k: int, n: int) -> int:
    MOD = 10**9 + 7
    s = min(k, n)
    t = max(k, n)

    num = denum = 1
    for i in range(1, s + 1):
        num = num * i % MOD
        denum = denum * ((t + i) % MOD) % MOD

    return pow(denum, MOD - 2, MOD) * num % MOD
```

```python
print(solve(73923861, 1, 1))                 # 500000004
print(solve(434360414, 2, 2))                # 166666668
print(solve(462964776, 90380, 234618688))    # 358688676
```
