```python
import sys
import numpy as np
from sklearn.svm import LinearSVC

def find_vector(X, y, m):
    if len(set(y)) == 1:
        return [y[0]] * m

    model = LinearSVC()
    model.fit(X, y)
    return model.coef_[0]

def main():
    n, m = map(int, input().split())
    X = np.zeros((n, m), float)
    y = np.zeros(n, int)
    for i in range(n):
        x = list(map(float, input().split()))
        X[i] = x[:-1]
        y[i] = int(x[-1] == 1)
    print(*find_vector(X, y, m))
 
if __name__ == '__main__':
    main()
```
