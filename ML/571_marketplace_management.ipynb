```python
import sys
import numpy as np


def compute_unmet_requests(predicted_requests, s, m):
    unmet_requests_total = np.zeros(m)
    for requests in predicted_requests:
        requests = np.array(requests)
        total_requests = np.sum(requests)
        if total_requests <= s:
            unmet_requests = np.zeros(m)
        else:
            unmet_requests = requests - (s / m)
            unmet_requests = np.maximum(unmet_requests, 0)
        unmet_requests_total += unmet_requests
    return unmet_requests_total / len(predicted_requests)
    

def main():
    m, s = map(int, input().split())
    n = int(input())
    historical_data = []

    for _ in range(n):
        line = list(map(int, input().split()))
        week, day, hour = line[:3]
        requests = line[3:]
        historical_data.append((week, day, hour, requests))

    q = int(input())
    queries = []
    for _ in range(q):
        line = list(map(int, input().split()))
        query_type = line[0]
        week, day, hour = line[1:4]
        if query_type == 1:
            requests = line[4:]
            historical_data.append((week, day, hour, requests))
        elif query_type == 2:
            queries.append((week, day, hour))
    
    data_by_time = {}
    for week, day, hour, requests in historical_data:
        key = (day, hour)
        if key not in data_by_time:
            data_by_time[key] = []
        data_by_time[key].append(requests)
    
    predicted_requests = []
    for week, day, hour in queries:
        key = (day, hour)
        if key in data_by_time:
            mean_requests = np.mean(data_by_time[key], axis=0)
        else:
            mean_requests = [0] * m
        predicted_requests.append(mean_requests)
    unmet_requests_avg = compute_unmet_requests(predicted_requests, s, m)
    print(" ".join(f"{x:.14f}" for x in unmet_requests_avg))


if __name__ == '__main__':
    main()
```
