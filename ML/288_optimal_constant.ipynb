```python
import sys


def optimal_constants(n, data):
    mse_constant = sum(data) / n
    
    sorted_data = sorted(data)
    if n % 2 == 1:
        mae_constant = sorted_data[n // 2]
    else:
        mae_constant = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2

    weighted_data = sorted((y, 1 / y) for y in data)
    total_weight = sum(weight for _, weight in weighted_data)
    cumulative_weight = 0
    for y, weight in weighted_data:
        cumulative_weight += weight
        if cumulative_weight >= total_weight / 2:
            mape_constant = y
            break
    
    return mse_constant, mae_constant, mape_constant


def main():
    n = int(input())
    data = [int(input()) for _ in range(n)]
    mse, mae, mape = optimal_constants(n, data)
    print(f"{mse:.6f}")
    print(f"{mae:.6f}")
    print(f"{mape:.6f}")


if __name__ == '__main__':
    main()
```
