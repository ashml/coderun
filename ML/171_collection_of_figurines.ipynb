```python
import sys


def find_min_price(n, k, prices):
    wanted = set(range(1, k + 1))
    current_set = set()
    current_count = {}
    min_price = float('inf')
    price = 0
    left = 0

    for right in range(n):
        price += prices[right]
        if prices[right] in wanted:
            if prices[right] not in current_count:
                current_count[prices[right]] = 0
            current_count[prices[right]] += 1
            current_set.add(prices[right])

        while len(current_set) == len(wanted):
            min_price = min(min_price, price)
            price -= prices[left]
            if prices[left] in current_count:
                current_count[prices[left]] -= 1
                if current_count[prices[left]] == 0:
                    current_set.discard(prices[left])
            left += 1

    return min_price if min_price != float('inf') else -1


def main():
    n, k = map(int, input().split())
    prices = list(map(int, input().split()))
    print(find_min_price(n, k, prices))


if __name__ == '__main__':
    main()
```
