```python
import sys


def main():
    step = 1.0
    delta = 1e-4
    q_count = 0

    line = sys.stdin.readline().split()
    x, y, h = float(line[0]), float(line[1]), float(line[2])
    while q_count < 200:
        print(f"{x + delta:.10f} {y:.10f}")
        sys.stdout.flush()
        res = sys.stdin.readline().split()
        hx = float(res[1])
        if res[0] == '+': 
            break
        
        print(f"{x:.10f} {y + delta:.10f}")
        sys.stdout.flush()
        res = sys.stdin.readline().split()
        hy = float(res[1])
        if res[0] == '+': 
            break
        
        q_count += 2
        
        gx, gy = (hx - h) / delta, (hy - h) / delta
        mag = (gx**2 + gy**2)**0.5
        if mag > 1e-12:
            dx, dy = gx / mag, gy / mag
        else:
            dx, dy = 1.0, 0.0

        while q_count < 200:
            nx, ny = x + dx * step, y + dy * step
            nx, ny = max(-50.0, min(100.0, nx)), max(-50.0, min(100.0, ny))
            
            print(f"{nx:.10f} {ny:.10f}")
            sys.stdout.flush()
            res = sys.stdin.readline().split()
            nh = float(res[1])
            q_count += 1
            
            if res[0] == '+': 
                sys.exit(0)
            
            if nh > h:
                x, y, h = nx, ny, nh
                step *= 1.5
                break
            else:
                step *= 0.5
                if step < 1e-7: 
                    break


if __name__ == "__main__":
    main()
```
