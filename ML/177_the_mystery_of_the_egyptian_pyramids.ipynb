```python
import sys
import numpy as np
from scipy.optimize import minimize


def model(x, a, b, c, d):
    return a * np.tan(x) + (b * np.sin(x) + c * np.cos(x))**2 + d * np.sqrt(x)


def loss(params, x_data, f_data):
    a, b, c, d = params
    predictions = model(x_data, a, b, c, d)
    return np.mean((predictions - f_data)**2)


def main():
    n = int(input())
    data = [tuple(map(float, input().split())) for _ in range(n)]
    
    x_data, f_data = zip(*data)
    x_data = np.array(x_data)
    f_data = np.array(f_data)

    bounds = [(0, None), (0, None), (0, None), (0, None)]
    initial_guess = [1, 1, 1, 1]
    result = minimize(loss, initial_guess, bounds=bounds, args=(x_data, f_data))
    a, b, c, d = result.x
    print(f"{a:.2f} {b:.2f} {c:.2f} {d:.2f}")


if __name__ == '__main__':
    main()
```
