```python
import sys
import heapq


def main():
    n, m = map(int, input().split())
    b = [int(input()) for _ in range(n)]

    by_type = [[] for _ in range(m)]
    for i, t in enumerate(b):
        by_type[t].append(i)

    next_pos = [0] * m
    heap = []
    for t in range(m):
        if by_type[t]:
            heapq.heappush(heap, (by_type[t][0], t))

    res = []
    prev = -1
    while heap:
        first_idx, first_t = heapq.heappop(heap)
        if first_t != prev:
            res.append(first_idx)
            next_pos[first_t] += 1
            if next_pos[first_t] < len(by_type[first_t]):
                heapq.heappush(heap, (by_type[first_t][next_pos[first_t]], first_t))
            prev = first_t
        elif not heap:
            break
        else:
            second_idx, second_t = heapq.heappop(heap)
            res.append(second_idx)
            next_pos[second_t] += 1
            if next_pos[second_t] < len(by_type[second_t]):
                heapq.heappush(heap, (by_type[second_t][next_pos[second_t]], second_t))
            prev = second_t
            heapq.heappush(heap, (first_idx, first_t))

    print(*res)


if __name__ == '__main__':
    main()
```
