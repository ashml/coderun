```python
import sys


def main():
    A = []
    B = []
    n, m = [int(num) for num in input().split()]
    for _ in range(n):
        A.append([int(num) for num in input().split()])
    k = int(input())
    for _ in range(k):
        B.append([int(num) for num in input().split()])
    rows = n - k + 1
    cols = m - k + 1
    conv = [[0 for _ in range(cols)] for _ in range(rows)]
    for i in range(rows):
        for j in range(cols):
            curr_sum = 0
            for t in range(k):
                for r in range(k):
                    curr_sum += A[i + t][j + r] * B[t][r]
            conv[i][j] = curr_sum
    for row in conv:
        print(' '.join([str(el) for el in row]))


if __name__ == '__main__':
    main()
```
