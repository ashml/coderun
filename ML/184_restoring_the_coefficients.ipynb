```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit


def func(x, a, b, c):
    return ((a * np.sin(x) + b * np.log(x)) ** 2) + (c * x**2)


data = pd.read_csv('data/184_restoring_the_coefficients/data.csv', header=None, names=['x', 'f_x'])
initial_guess = [1.0, 1.0, 1.0]
popt, _ = curve_fit(func, data['x'], data['f_x'], p0=initial_guess)
a, b, c = popt
print(f"{a:.2f} {b:.2f} {c:.2f}")
```
