```python
import sys


def phi(n):
    result = n
    p = 2
    while p * p <= n:
        if n % p == 0:
            while n % p == 0:
                n //= p
            result -= result // p
        p += 1
    if n > 1:
        result -= result // n
    return result


def main():
    n = int(input())
    print(phi(n))


if __name__ == "__main__":
    main()
```
