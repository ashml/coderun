```python
pigeons = 500
cats = 500

timur = {'tp': 0.8 * pigeons, 'fn': 0.2 * pigeons, 'fp': 0.4 * cats, 'tn': 0.6 * cats}
kirill = {'tp': 0.6 * pigeons, 'fn': 0.4 * pigeons, 'fp': 0.2 * cats, 'tn': 0.8 * cats}
```

```python
def f1_score(tp, fp, fn):
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    
    if precision + recall == 0:
        return 0
    return 2 * (precision * recall) / (precision + recall)
```

```python
max_tp = cats + pigeons - kirill['fp'] - timur['fn'] - timur['tn']
f1 = f1_score(max_tp, kirill['fp'], timur['fn'])
print(f"F1-мера: {f1:.6f}")
```
