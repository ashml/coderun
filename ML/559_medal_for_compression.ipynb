```python
import sys
import numpy as np
from sklearn.decomposition import PCA


def main():
    first_line = input().split()
    n, m, p = int(first_line[0]), int(first_line[1]), float(first_line[2])

    data = np.zeros((n, m + 1))
    for i in range(n):
        row = list(map(float, input().strip().split()))
        data[i, :] = row   

    v = int((2/3) * m)
    pca = PCA(n_components=v)
    X_reduced = pca.fit_transform(data[:, :-1])
    
    print(v)
    for i in range(n):
        print(' '.join(map(str, X_reduced[i])))


if __name__ == '__main__':
    main()
```
