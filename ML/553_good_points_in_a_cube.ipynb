```python
import numpy as np
from scipy.integrate import tplquad


R = 0.75
a = 1
x0, y0, z0 = 0.5, 0.5, 0.5


def integrand(z, y, x):
    return ((x - x0)**2 + (y - y0)**2 + (z - z0)**2 < R**2)
    
def bounds_x():
    return [0, a]

def bounds_y(x):
    return [0, a]

def bounds_z(x, y):
    return [0, a]


volume, error = tplquad(
    integrand, 
    0, a, 
    lambda x: 0, lambda x: a, 
    lambda x, y: 0, lambda x, y: a, 
    epsabs=1e-5, epsrel=1e-5
)

print(f"Объём сферы внутри куба: {volume}")
print(round(1 - volume, 5))
```
