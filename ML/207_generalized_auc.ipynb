```python
'''TOO SLOW'''
n = int(input())
pairs = [tuple(map(float, input().split())) for _ in range(n)]

numerator = 0
denominator = 0

for i in range(n):
    for j in range(n):
        if pairs[i][0] > pairs[j][0]:
            denominator += 1
            if pairs[i][1] > pairs[j][1]:
                numerator += 1
            elif pairs[i][1] == pairs[j][1]:
                numerator += 0.5


generalized_auc = numerator / denominator if denominator > 0 else 0
print(f"{generalized_auc:.6f}")
```

```python
import sys


class Fenwick:
    def __init__(self, n):
        self.n = n
        self.t = [0] * (n + 1)

    def add(self, i, v):
        while i <= self.n:
            self.t[i] += v
            i += i & -i

    def sum(self, i):
        s = 0
        while i > 0:
            s += self.t[i]
            i -= i & -i
        return s


def main():
    n = int(input())
    pairs = [tuple(map(float, input().split())) for _ in range(n)]
    uniq_y = set([el[1] for el in pairs])

    m = len(uniq_y)
    rank = {v: i + 1 for i, v in enumerate(sorted(uniq_y))}

    pairs.sort()
    bit = Fenwick(m)
    freq = [0] * (m + 1)

    numerator = 0
    denominator = 0
    total_seen = 0

    i = 0
    while i < n:
        curr = pairs[i][0]
        group = []
        while i < n and pairs[i][0] == curr:
            group.append(pairs[i])
            i += 1

        for t, y in group:
            rnk = rank[y]
            less = bit.sum(rnk - 1)
            equal = freq[rnk]
            numerator += less + equal * 0.5
            denominator += total_seen

        for t, y in group:
            rnk = rank[y]
            bit.add(rnk, 1)
            freq[rnk] += 1
            total_seen += 1

    auc = numerator / denominator
    print(f"{auc:.6f}")


if __name__ == "__main__":
    main()
```
