```python
import sys
from collections import defaultdict

def build_frontier(starter, rules, max_steps, max_len):
    fronts = [set() for _ in range(max_steps + 1)]
    fronts[0].add(starter)
    for depth in range(1, max_steps + 1):
        nxt = fronts[depth]
        for s in fronts[depth - 1]:
            for a, b in rules.items():
                start = 0
                while True:
                    i = s.find(a, start)
                    if i < 0: 
                        break
                    for el in b:
                        new = s[:i] + el + s[i + len(a):]
                        if len(new) <= max_len:
                            nxt.add(new)
                    start = i + 1
    return fronts

def main():
    data = sys.stdin.read().split()
    start, target = data[0], data[1]
    n = int(data[2])
    raw = data[3:]
    forward = defaultdict(list)
    backward = defaultdict(list)
    max_diff = 0
    for i in range(0, 2*n, 2):
        a, b = raw[i], raw[i+1]
        forward[a].append(b)
        backward[b].append(a)
        max_diff = max(max_diff, abs(len(b) - len(a)))

    max_len = max(len(start), len(target)) + 4 * max_diff

    F = build_frontier(start, forward, 2, max_len)
    B = build_frontier(target, backward, 2, max_len)
    ans = None
    for i in range(3):
        for j in range(3):
            if i + j <= 4 and (F[i] & B[j]):
                d = i + j
                if ans is None or d < ans:
                    ans = d
    print(ans if ans is not None else -1)

if __name__ == "__main__":
    main()
```
