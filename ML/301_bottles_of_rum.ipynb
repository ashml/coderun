```python
import sys
from collections import Counter
from math import comb


def main():
    alpha = input()
    n = len(alpha)
    cnt = Counter(alpha)
    ways = [0] * (n + 1)
    ways[0] = 1
    curr = 0
    for m in cnt.values():
        for j in range(curr, -1, -1):
            way = ways[j]
            if way == 0:
                continue
            for t in range(1, m + 1):
                ways[j + t] += way * comb(j + t, t)
        curr += m
    print(sum(ways) - 1)


if __name__ == '__main__':
    main()
```
