```python
import sys
import math


def main():
    n, k = map(int, input().split())
    data = []
    for _ in range(n):
        line = list(map(float, input().split()))
        x = line[:k]
        a, b = line[k], line[k+1]
        data.append((a, b))
    
    t = [a / b for a, b in data]
    w = [b for _, b in data]
    sum_w = sum(w)
    p_mse = sum(w[i] * t[i] for i in range(n)) / sum_w
    p_msle = math.exp(sum(w[i] * math.log(1 + t[i]) for i in range(n)) / sum_w) - 1
    p_logloss = p_mse
    print(f"{p_mse:.6f} {p_msle:.6f} {p_logloss:.6f}")


if __name__ == '__main__':
    main()
```
