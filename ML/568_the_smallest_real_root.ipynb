```python
import sys
import numpy as np


def find_min_real_root(n, A, B):
    I = np.eye(n)
    Z = np.zeros_like(I)
    block_matrix = np.block([[Z, I], [-B, -A]])
    
    eigenvalues, _ = np.linalg.eig(block_matrix)
    real_eigenvalues = np.real(eigenvalues)
    min_root = np.min(real_eigenvalues)
    return np.round(min_root, 4)
    

def main():
    n = int(input())
    A = np.array([[float(el) for el in input().split()] for _ in range(n)])
    B = np.array([[float(el) for el in input().split()] for _ in range(n)])
    print(find_min_real_root(n, A, B))


if __name__ == '__main__':
    main()
```
