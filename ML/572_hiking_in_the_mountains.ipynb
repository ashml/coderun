```python
import sys


def max_stay_time(n, s, arrival_times):
    if n <= s:
        return 'INF'
    arrival_times.sort()
    max_time = float('inf')
    for i in range(s, n):
        if arrival_times[i] == arrival_times[i - s]:
            return 'Impossible'
        max_time = min(arrival_times[i] - arrival_times[i - s], max_time)
    return max_time


def main():
    n, s = [int(el) for el in input().split()]
    times = [int(input()) for _ in range(n)]
    print(max_stay_time(n, s, times))


if __name__ == '__main__':
    main()
```
