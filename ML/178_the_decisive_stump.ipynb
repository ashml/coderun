```python
import sys


def main():
    n = int(input())
    points = [tuple(map(int, input().split())) for _ in range(n)]

    points.sort(key=lambda p: p[0])
    xs = [x for x, _ in points]
    ys = [y for _, y in points]

    total_sum = sum(ys)
    prefix_sum = [0] * (n + 1)
    for i in range(n):
        prefix_sum[i + 1] = prefix_sum[i] + ys[i]

    best_value = (total_sum * total_sum) / n
    best_split = 0

    for i in range(1, n):
        if xs[i - 1] == xs[i]:
            continue
        left_size = i
        left_sum = prefix_sum[left_size]
        right_sum = total_sum - left_sum
        value = (left_sum ** 2) / left_size + (right_sum ** 2) / (n - left_size)

        if value > best_value:
            best_value = value
            best_split = left_size

    if best_split == 0:
        a = 0.0
        b = total_sum / n
        c = xs[0] - 1.0
    elif best_split == n:
        a = total_sum / n
        b = 0.0
        c = xs[-1] + 1.0
    else:
        left_sum = prefix_sum[best_split]
        right_sum = total_sum - left_sum
        a = left_sum / best_split
        b = right_sum / (n - best_split)
        c = (xs[best_split - 1] + xs[best_split]) / 2.0

    print(f"{a:.6f} {b:.6f} {c:.6f}")


if __name__ == '__main__':
    main()
```
