```python
import sys
import numpy as np


def main():
    A = []
    C = []
    coeffs = []
    consts = []
    n, m, k = [int(num) for num in input().split()]
    
    for _ in range(n):
        A.append([int(num) for num in input().split()])
        
    for _ in range(n - k + 1):
        C.append([int(num) for num in input().split()])

    for i in range(n - k + 1):
        for j in range(m - k + 1):
            equation = []
            for t in range(k):
                for l in range(k):
                    equation.append(A[i + t][j + l])
            coeffs.append(equation)
            consts.append(C[i][j])
        
    arr, *_ = np.linalg.lstsq(np.array(coeffs), np.array(consts), rcond=None)
    B = arr.reshape((k, k)).round().astype(int)
        
    for row in B:
        print(' '.join([str(el) for el in row]))


if __name__ == '__main__':
    main()
```
