```python
import sys
from math import gcd


MOD = 1000000007


def modinv(a, p):
    return pow(a, p - 2, p)


def get_prime_factors_up_to(n):
    sieve = list(range(n + 1))
    for i in range(2, int(n**0.5) + 1):
        if sieve[i] == i:
            for j in range(i * i, n + 1, i):
                if sieve[j] == j:
                    sieve[j] = i
    primes = set()
    for i in range(2, n + 1):
        x = i
        while x != 1:
            primes.add(sieve[x])
            x //= sieve[x]
    return sorted(primes)


def count_square_products(m, k):
    primes = get_prime_factors_up_to(k)
    prime_index = {p: i for i, p in enumerate(primes)}
    num_primes = len(primes)
    
    dp = [0] * (1 << num_primes)
    dp[0] = 1
    
    for _ in range(m):
        next_dp = [0] * (1 << num_primes)
        for i in range(k):
            fac = i + 1
            fac_bits = 0
            for p in primes:
                if fac % p == 0:
                    count = 0
                    while fac % p == 0:
                        fac //= p
                        count += 1
                    if count % 2 == 1:
                        fac_bits ^= (1 << prime_index[p])
            for mask in range(1 << num_primes):
                next_dp[mask ^ fac_bits] = (next_dp[mask ^ fac_bits] + dp[mask]) % MOD
        dp = next_dp
    
    total_outcomes = pow(k, m, MOD)
    winning_outcomes = dp[0]
    
    p = winning_outcomes
    q = total_outcomes
    g = gcd(p, q)
    p //= g
    q //= g
    
    return (p * modinv(q, MOD)) % MOD


def main():
    m, k = [int(el) for el in input().split()]
    result = count_square_products(m, k)
    print(result)


if __name__ == "__main__":
    main()
```
