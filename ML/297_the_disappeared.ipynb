```python
import sys


def rotate_90(image):
    return [''.join(row) for row in zip(*image[::-1])]


def find_object(n1, m1, cat, n2, m2, frame):
    rotations = [cat]
    for _ in range(3):
        rotations.append(rotate_90(rotations[-1]))

    for rotation in rotations:
        h_cat = len(rotation)
        w_cat = len(rotation[0])
        if h_cat > n2 or w_cat > m2:
            continue

        for i in range(n2 - h_cat + 1):
            start_col = 0
            while True:
                idx = frame[i].find(rotation[0], start_col)
                if idx == -1:
                    break

                match = True
                for k in range(1, h_cat):
                    if frame[i + k][idx : idx + w_cat] != rotation[k]:
                        match = False
                        break
                
                if match:
                    return 'Yes'

                start_col = idx + 1
    return 'No'


def main():
    n1, m1 = map(int, input().split())
    cat = [input() for _ in range(n1)]
    n2, m2 = map(int, input().split())
    frame = [input() for _ in range(n2)]
    print(find_object(n1, m1, cat, n2, m2, frame))


if __name__ == '__main__':
    main()
```
