```python
import pandas as pd
```

```python
data_path = 'data/312_innocent_and_friends/'
users = pd.read_csv(data_path + 'users.csv')
orgs = pd.read_csv(data_path + 'organisations.csv', usecols=['org_id', 'city'])
reviews = pd.read_csv(data_path + 'reviews.csv', usecols=['user_id', 'org_id', 'rating'])
test_users = pd.read_csv(data_path + 'test_users.csv')
```

```python
def get_recommendation(home_city):
    if home_city == 'msk':
        return top_spb_str
    else:
        return top_msk_str

good_reviews = reviews[reviews['rating'] >= 4]
joined = good_reviews.merge(orgs, on='org_id', how='inner')
org_popularity = joined.groupby(['city', 'org_id']).size().reset_index(name='review_count')
org_popularity = org_popularity.sort_values(by=['city', 'review_count'], ascending=[True, False])

top_msk_list = org_popularity[org_popularity['city'] == 'msk'].head(100)['org_id'].tolist()
top_spb_list = org_popularity[org_popularity['city'] == 'spb'].head(100)['org_id'].tolist()

top_msk_str = str(top_msk_list)
top_spb_str = str(top_spb_list)

test_users_merged = test_users.merge(users, on='user_id', how='left')
test_users_merged['target'] = test_users_merged['city'].apply(get_recommendation)

output = test_users_merged[['user_id', 'target']]
output.to_csv(data_path + 'result.csv', index=False)

print("Файл result.csv успешно сформирован.")
```
