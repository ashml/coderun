```python
import sys
from math import gcd
from collections import defaultdict


def main():
    n = int(input())
    p = input().split()
    v = input().split()

    p_idx = defaultdict(int)
    v_idx = defaultdict(int)
    pairs = defaultdict(int)
    
    P = 0
    for i in range(len(p)):
        P += i - (p_idx[p[i]] - pairs[(p[i], v[i])]) - (v_idx[v[i]] - pairs[(p[i], v[i])])

        pairs[(p[i], v[i])] += 1
        p_idx[p[i]] += 1
        v_idx[v[i]] += 1

    Q = int(n * (n - 1) * 0.5)
    common_divisor = gcd(P, Q)
    P //= common_divisor
    Q //= common_divisor
    print(f"{P}/{Q}")


if __name__ == '__main__':
    main()
```
