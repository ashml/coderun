```python
import sys
import hashlib
import math


M = 2 ** 15
bitset = bytearray(M // 8)

def set_bit(pos):
    byte_i = pos >> 3
    bit_i  = pos & 7
    bitset[byte_i] |= (1 << bit_i)

def is_zero(pos):
    byte_i = pos >> 3
    bit_i  = pos & 7
    return (bitset[byte_i] & (1 << bit_i)) == 0

def main():
    n = int(input())
    for _ in range(n):
        h = hashlib.sha1(input().strip().encode()).digest()
        idx = int.from_bytes(h[:8], 'big', signed=False) % M
        set_bit(idx)

    V = sum(int(is_zero(i)) for i in range(M))
    print(-M * math.log(V / M))

if __name__ == "__main__":
    main()
```
