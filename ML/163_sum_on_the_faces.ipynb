```python
import sys
from collections import Counter


def main():
    dice_faces = list(map(int, input().split()))
    num_rolls = int(input())

    face_counts = Counter(dice_faces)
    expected_first_roll = sum(dice_faces) / 6
    weighted_sum_for_repeats = 0
    
    for face_value, count in face_counts.items():
        weighted_sum_for_repeats += count * (6 - count) * face_value

    expected_value_for_remaining_rolls = weighted_sum_for_repeats / 36.0
    total_expected_sum = expected_first_roll + (num_rolls - 1) * expected_value_for_remaining_rolls
    print("{0:.10f}".format(total_expected_sum))


if __name__ == "__main__":
    main()
```
