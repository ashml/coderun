```python
import sys
import math


def calculate_ropeway_length(n, peaks):
    total = 0.0
    px, py = 0.0, 0.0
    for x, y in peaks:
        total += math.hypot(x - px, y - py)
        px, py = x, y

    nxt = [-1] * n
    stack = []
    for i, (x, y) in enumerate(peaks):
        while stack and peaks[stack[-1]][1] <= y:
            idx = stack.pop()
            nxt[idx] = i
        stack.append(i)

    for i in range(n):
        j = nxt[i]
        if j != -1 and j > i + 1:
            xi, yi = peaks[i]
            xj, yj = peaks[j]
            total += math.hypot(xj - xi, yj - yi)

    return total


def main():
    n = int(input())
    peaks = [tuple(map(int, input().split())) for _ in range(n)]
    result = calculate_ropeway_length(n, peaks)
    print(f"{result:.12f}")


if __name__ == "__main__":
    main()
```
