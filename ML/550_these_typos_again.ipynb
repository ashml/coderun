```python
import sys


def get_levenshtein_distance(s, t, k):
    n, m = len(s), len(t)
    if n > m:
        s, t = t, s
        n, m = m, n

    if k > max(n, m):
        return 0
    if m - n > k:
        return k + 1

    p = [0] * (n + 1)
    d = [0] * (n + 1)
    boundary = min(n, k) + 1
    for i in range(boundary):
        p[i] = i
    p[boundary:] = [float('inf')] * (n + 1 - boundary)
    d[:] = [float('inf')] * (n + 1)
    for j in range(1, m + 1):
        t_j = t[j - 1]
        d[0] = j
        min_index = max(1, j - k)
        max_index = min(n, j + k)
        if min_index > max_index:
            return k + 1
        if min_index > 1:
            d[min_index - 1] = float('inf')
        for i in range(min_index, max_index + 1):
            if s[i - 1] == t_j:
                d[i] = p[i - 1]
            else:
                d[i] = 1 + min(d[i - 1], p[i], p[i - 1])
        p, d = d, p
    return p[n]


def main():
    n = int(input())
    for _ in range(n):
        k = int(input())
        s = input()
        t = input()
        distance = get_levenshtein_distance(s, t, k)
        print('Yes') if distance <= k else print('No')


if __name__ == '__main__':
    main()
```
