```python
import numpy as np
import math
from sklearn.cluster import KMeans


file_path = 'data/556_in_pursuit_of_profit/easy_pursuit_of_profit_input.txt'
output_path = 'data/556_in_pursuit_of_profit/output.txt'
expenditures = 0
with open(file_path, 'r') as file:
    k, n, c_cost, C = map(int, file.readline().strip().split())
    houses = []
    for _ in range(n):
        x, y = map(int, file.readline().strip().split())
        houses.append((x, y))

houses = np.array(houses)
kmeans = KMeans(n_clusters=k, n_init=20, random_state=42).fit(houses)
clusters = kmeans.predict(houses)
centers = kmeans.cluster_centers_

for i in range(k):
    cluster_houses = houses[clusters == i]
    center = centers[i]
    distances = np.linalg.norm(cluster_houses - center, axis=1)
    expenditures += c_cost * sum((distances**0.25 + 1) / distances.shape[0])

C_found = C - expenditures

with open(output_path, 'w') as file:
    file.write(f"{C_found:.9f}\n")
    for i in range(k):
        file.write(f"{centers[i][0]:.9f} {centers[i][1]:.9f}\n")
    
    for i in range(k):
        cluster_houses_indices = np.where(clusters == i)[0]
        file.write(" ".join(map(str, cluster_houses_indices.tolist())) + "\n")
```
