```python
import sys


def check_str(text, stops):
    for word in stops:
        if text.find(word) != -1:
            return 'DELETE'
    return 'KEEP'


def main():
    n, m = [int(el) for el in input().split()]
    stops = {input().lower() for _ in range(n)}
    for _ in range(m):
        text = input().lower()
        print(check_str(text, stops))


if __name__ == '__main__':
    main()
```
