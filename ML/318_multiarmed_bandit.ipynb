```python
import sys
import random


def main():
    while True:
        n, m = map(int, input().split())
        if n == 0 and m == 0:
            return
        alpha_in, beta_in = map(float, input().split())
        successes = [0] * m
        failures = [0] * m
        
        for _ in range(n):
            best_arm = 0
            max_val = -1.0
            for i in range(m):
                sample = random.betavariate(alpha_in + successes[i], beta_in + failures[i])
                if sample > max_val:
                    max_val = sample
                    best_arm = i
            print(best_arm + 1)
            sys.stdout.flush()
            result = int(sys.stdin.readline())
            if result == 1:
                successes[best_arm] += 1
            else:
                failures[best_arm] += 1


if __name__ == "__main__":
    main()
```
