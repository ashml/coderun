```python
import sys
from collections import defaultdict


def main():
    words = defaultdict(int)
    substrings = defaultdict(list)
    pairs = []
    for i in range(1, int(input()) + 1):
        s = input()
        for j in substrings[s]:
            pairs.append((i, j))
            pairs.append((j, i))
        for n in range(len(s) // 2 - (len(s) + 1) % 2):
            if s[:n + 1] == s[-1 - n:]:
                substr = s[n + 1:-1 - n]
                if substr in words:
                    pairs.append((i, words[substr]))
                    pairs.append((words[substr], i))
                substrings[substr].append(i)
        words[s] = i
        
    for el in sorted(pairs):
        print(*el)


if __name__ == '__main__':
    main()
```
