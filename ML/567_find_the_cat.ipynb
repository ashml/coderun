```python
import sys
sys.setrecursionlimit(10000)


def count_and_label_cats(matrix):
    M = len(matrix)
    N = len(matrix[0])
    def dfs(x, y, label):
        if x < 0 or x >= M or y < 0 or y >= N or matrix[x][y] != 1:
            return
        matrix[x][y] = label
        dfs(x - 1, y, label)
        dfs(x + 1, y, label)
        dfs(x, y - 1, label)
        dfs(x, y + 1, label)

    label = 0
    for i in range(M):
        for j in range(N):
            if matrix[i][j] == 1:
                label += 1
                dfs(i, j, label + 1)
    return label, matrix
    

def main():
    matrix = []
    for line in sys.stdin.readlines():
        row = list(map(int, line.split()))
        matrix.append(row)
        
    number_of_cats, labeled_matrix = count_and_label_cats(matrix)
    
    print(number_of_cats)
    for row in labeled_matrix:
        print(' '.join(map(str, row)))


if __name__ == '__main__':
    main()
```
