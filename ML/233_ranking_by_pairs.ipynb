```python
import sys
import numpy as np
from scipy.optimize import minimize

def loss_and_grad(f):
    loss = 0.0
    grad = np.zeros_like(f)

    for i, j in np.array(pairs):
        d = f[j] - f[i]
        p = 1 / (1 + np.exp(d))
        loss -= np.log(p)

        g = 1 - p
        grad[i] -= g
        grad[j] += g

    return loss, grad


n, m = map(int, input().split())
pairs = [tuple(int(x) - 1 for x in input().split()) for _ in range(m)]

res = minimize(
    fun=lambda x: loss_and_grad(x)[0],
    x0=np.zeros(n),
    jac=lambda x: loss_and_grad(x)[1],
    method="L-BFGS-B",
    options={"maxiter": 50}
)

f = res.x
ranking = np.argsort(-f) + 1
print('\n'.join(map(str, ranking)))
```
