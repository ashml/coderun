```python
import sys
from collections import defaultdict


def create_rec(p, n, k, videos, ids):
    v_count = defaultdict(int)
    result = []
    for i in range(p):
        topic = videos[i]
        v_id = ids[i]
        if v_count[topic] < k:
            v_count[topic] += 1
            result.append(f"{topic} #{v_id}")
        if len(result) == n:
            return result
    return result


def main():
    p, n, k = [int(el) for el in input().split()]
    videos = [input() for _ in range(p)]
    ids = [int(el) for el in input().split()]
    for el in create_rec(p, n, k, videos, ids):
        print(el)


if __name__ == '__main__':
    main()
```
