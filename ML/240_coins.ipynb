```python
import pandas as pd


input_file = 'data/240_coins/coins.in'
output_file = 'data/240_coins/answer.csv'

with open(input_file, 'r') as file:
    lines = file.readlines()

n = int(lines[0].strip())
data = [tuple(map(int, line.strip().split())) for line in lines[1:]]

df = pd.DataFrame(data, columns=['k', 'm'])
df['index'] = df.index
df['p'] = (df['m'] + 1) / (df['k'] + 2)
sorted_df = df.sort_values(by='p', ascending=True)
sorted_indices = sorted_df['index']
sorted_indices.to_csv(output_file, index=False, header=False)
```
