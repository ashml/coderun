```python
import sys
from collections import defaultdict, deque


def determine_quarantine(n, test_results, meeting_info):
    meetings = defaultdict(list)
    for employee_id, employee_meetings in enumerate(meeting_info):
        for meeting in employee_meetings:
            meetings[meeting].append(employee_id)
    sorted_meetings = sorted(meetings.keys())
    infected = set(i for i, result in enumerate(test_results) if result == 1)
    quarantine = set(infected)
    for meeting in sorted_meetings:
        participants = meetings[meeting]
        if any(p in infected for p in participants):
            for participant in participants:
                if participant not in infected:
                    infected.add(participant)
                    quarantine.add(participant)
    return [1 if i in quarantine else 0 for i in range(n)]


def main():
    n = int(input())
    test_results = list(map(int, input().split()))
    meeting_info = []
    for _ in range(n):
        line = list(map(int, input().split()))
        meeting_info.append(line[1:])
    result = determine_quarantine(n, test_results, meeting_info)
    print(*result)


if __name__ == '__main__':
    main()
```
