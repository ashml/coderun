```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


train_path = 'data/200_warm_up/train.tsv'
test_path = 'data/200_warm_up/test.tsv'
output_path = 'data/200_warm_up/answer.tsv'


train_data = pd.read_csv(train_path, sep='\t', header=None)
test_data = pd.read_csv(test_path, sep='\t', header=None)

X_train = train_data.iloc[:, :-1].values
y_train = train_data.iloc[:, -1].values

X_test = test_data.values

model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
train_error = np.max(np.abs(y_train - y_train_pred))
print(f"Максимальная абсолютная ошибка на обучении: {train_error}")

assert train_error <= 1e-8, "Точность модели на обучающем наборе недостаточна!"

y_test_pred = model.predict(X_test)
pd.DataFrame(y_test_pred).to_csv(output_path, header=False, index=False, float_format="%.8f")
print(f"Предсказания сохранены в файл: {output_path}")
```
