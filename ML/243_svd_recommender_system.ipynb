```python
import numpy as np
import sys

def main():   
    k, U, M, D, T = map(int, input().split())
    train_data = [list(map(int, input().split())) for _ in range(D)]
    test_data = [list(map(int, input().split())) for _ in range(T)]
        
    total_rating = 0.0
    for (u, m, r) in train_data:
        total_rating += r
    mu = total_rating / D

    d = 20
    b_u = np.zeros(U)
    b_m = np.zeros(M)
    p_u = np.random.normal(0, 0.1, (U, d))
    q_m = np.random.normal(0, 0.1, (M, d))
    
    n_epochs = 15
    learning_rate = 0.03
    reg = 0.1
    
    for epoch in range(n_epochs):
        np.random.shuffle(train_data)
        for u, m, r in train_data:
            dot_product = np.dot(p_u[u], q_m[m])
            prediction = mu + b_u[u] + b_m[m] + dot_product
            error = r - prediction
            
            b_u[u] += learning_rate * (error - reg * b_u[u])
            b_m[m] += learning_rate * (error - reg * b_m[m])
            p_u[u] += learning_rate * (error * q_m[m] - reg * p_u[u])
            q_m[m] += learning_rate * (error * p_u[u] - reg * q_m[m])
                
    for u, m in test_data:
        prediction = mu + b_u[u] + b_m[m] + np.dot(p_u[u], q_m[m])
        print(prediction)

if __name__ == '__main__':
    main()
```
