```python
import sys


def main():
    header = sys.stdin.readline()
    
    logs = []
    for line in sys.stdin:
        parts = line.strip().split('\t')
        if len(parts) < 4:
            continue
        val = float(parts[3]) if parts[3] else 0.0
        logs.append((parts[1], parts[0], parts[2], val))

    logs.sort()

    all_months = set()
    user_first_date = {}
    daily_revenue = {}
    
    for ts, uid, action, val in logs:
        current_date = ts[:10]
        current_month = ts[:7]
        
        all_months.add(current_month)
        
        if uid not in user_first_date:
            user_first_date[uid] = current_date
            
        if action == 'confirmation':
            if user_first_date[uid] == current_date:
                daily_revenue[current_date] = daily_revenue.get(current_date, 0.0) + val

    sorted_months = sorted(list(all_months))
    
    if len(sorted_months) >= 3:
        candidate_months = sorted_months[1:-1]
        
        for month in candidate_months:
            relevant_days = sorted([d for d in daily_revenue if d.startswith(month)])
            
            if not relevant_days:
                continue
                
            best_day = None
            max_revenue = -1.0
            
            for day in relevant_days:
                revenue = daily_revenue[day]
                if revenue > max_revenue:
                    max_revenue = revenue
                    best_day = day
            
            if best_day:
                print(f"{best_day} {int(max_revenue)}")


if __name__ == '__main__':
    main()
```
