```python
import random


def solve_task():
    time_spent = 0
    while time_spent < 5:
        if random.random() < 0.4:
            distraction_time = random.uniform(0, 10)
            time_spent += distraction_time
        else:
            time_spent += 0.1
    return time_spent


def simulate(num_simulations=10000):
    total_tasks_completed = 0
    for _ in range(num_simulations):
        time_spent = 0
        tasks_completed = 0
        while time_spent <= 60:
            time_for_task = solve_task()
            time_spent += time_for_task
            if time_spent <= 60:
                tasks_completed += 1
        total_tasks_completed += tasks_completed
    return total_tasks_completed / num_simulations


simulated_tasks = simulate()
print(f"Среднее количество задач, решённых за час: {simulated_tasks:.2f}")
```

```python
from sympy import symbols, Eq, solve


E_T = symbols('E_T')
equation = Eq(0.6 * 5 + 0.4 * (2.5 + 5 + E_T), E_T)
expected_time_per_task = solve(equation, E_T)[0]
tasks_in_one_hour = 60 / expected_time_per_task
print(f"Количество задач, которое Алиса может решить за 60 минут: {tasks_in_one_hour:.0f}")
```
