```python
import sys


def check_loop(path):
    team = 0
    visited = set()
    while team != -1:
        if team in visited:
            return 'YES'
        visited.add(team)
        team = path[team]
    return 'NO'


def main():
    n = int(input())
    redir = dict(zip(range(n), [int(el) for el in input().split()]))
    print(check_loop(redir))
        

if __name__ == '__main__':
    main()
```
