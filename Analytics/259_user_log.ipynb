```python
import pandas as pd
import os


data_path = "data/259_user_log/"
log_file = os.path.join(data_path, "log.csv")
log_df = pd.read_csv(log_file, header=None, names=["id", "action"])

unique_ids = log_df["id"].unique()
result_df = pd.DataFrame(unique_ids, columns=["id"])

for i in range(1, 21):
    feature_file = os.path.join(data_path, f"feature_{i}.csv")
    feature_df = pd.read_csv(feature_file, header=None, names=["id", f"feature_{i}"])
    result_df = result_df.merge(feature_df, on="id", how="left")

result_df["non_empty_features"] = result_df.iloc[:, 1:].notnull().sum(axis=1)
users_with_50_percent_features = (result_df["non_empty_features"] >= 10).sum()

print(users_with_50_percent_features)
```
