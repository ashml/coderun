```python
import math


n1, x1 = 5807, 30
n2, x2 = 5829, 101
alpha = 0.01
z_critical = 2.57


p1 = x1 / n1
p2 = x2 / n2

p_pooled = (x1 + x2) / (n1 + n2)
q_pooled = 1 - p_pooled

se_z = math.sqrt(p_pooled * q_pooled * (1/n1 + 1/n2))

z_stat = (p1 - p2) / se_z

reject_null = z_stat < -z_critical
q1_answer = "yes" if reject_null else "no"

se_diff = math.sqrt(p1 * (1 - p1) / n1 + p2 * (1 - p2) / n2)
ci_lower = (p1 - p2) - z_critical * se_diff
ci_upper = (p1 - p2) + z_critical * se_diff

z_stat_rounded = round(z_stat, 3)
ci_lower_rounded = round(ci_lower, 3)
ci_upper_rounded = round(ci_upper, 3)

q1_answer, z_stat_rounded, (ci_lower_rounded, ci_upper_rounded)

output = f"""Q1 {q1_answer}
Q2 {z_stat_rounded}
Q3 {ci_lower_rounded} {ci_upper_rounded}"""

print(output)
```
