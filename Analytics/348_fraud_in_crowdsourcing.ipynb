```python
import random
import numpy as np


def simulate_block(Y, num_simulations=100000):
    N = 100 - Y
    block_count = 0
    control_tasks = [1]*Y + [0]*N

    for _ in range(num_simulations):
        random.shuffle(control_tasks)
        
        errors = 0
        blocked = False
        last_10 = []

        for i in range(25):
            task = control_tasks[i]
            if task == 0:
                errors += 1
                last_10.append(1)
            else:
                last_10.append(0)

            if len(last_10) > 10:
                last_10.pop(0)

            if i == 9:
                if errors > 3:
                    block_count += 1
                    blocked = True
                    break
            elif i >= 14 and (i - 14) % 5 == 0:
                window_errors = sum(last_10)
                if window_errors > 3:
                    block_count += 1
                    blocked = True
                    break


    probability = block_count / num_simulations
    return probability


def find_max_Y(target_probability=0.8, num_simulations=100000):
    for Y in range(50, 101):
        prob = simulate_block(Y, num_simulations)
        print(f"Y = {Y}, Probability of blocking = {prob*100:.2f}%")
        if prob < target_probability:
            max_Y = Y - 1
            return max_Y
    return 100


if __name__ == "__main__":
    max_Y = find_max_Y()
    print(f"\nМаксимальное число заданий с ответом «Да»: {max_Y}")
```
