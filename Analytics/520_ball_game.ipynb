```python
import sys
from fractions import Fraction


def probability_game_ends(n):
    numerator = (n + 4) * (n + 3)
    denominator = (n + 6) * (n + 1) * 2
    fraction_result = Fraction(denominator - numerator, denominator)
    num = fraction_result.numerator
    denom = fraction_result.denominator
    return num, denom


def main():
    n = int(input())
    p, q = probability_game_ends(n)
    print(f"{p} {q}")


if __name__ == '__main__':
    main()
```
