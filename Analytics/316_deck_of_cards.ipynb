```python
from math import comb
from itertools import combinations_with_replacement


card_values = {2: 4, 3: 4, 4: 4, 5: 4, 6: 4, 7: 4, 8: 4, 9: 4, 10: 4, 11: 4, 12: 4, 13: 4, 14: 4}
total_cards = 52
hand_size = 6
target_sum = 21
card_nums = list(card_values.keys())


def find_valid_combinations():
    valid_combinations = []
    for counts in combinations_with_replacement(card_nums, hand_size):
        if sum(counts) == target_sum:
            counts_dict = {value: counts.count(value) for value in set(counts)}
            if all(counts_dict[value] <= card_values[value] for value in counts_dict):
                valid_combinations.append(counts_dict)
    return valid_combinations


def calculate_ways(combinations):
    total_ways = 0
    for combo in combinations:
        ways = 1
        for value in combo:
            ways *= comb(card_values[value], combo[value])
        total_ways += ways
    return total_ways


valid_combinations = find_valid_combinations()
favorable_ways = calculate_ways(valid_combinations)
total_ways = comb(total_cards, hand_size)
probability = favorable_ways / total_ways
print(f"{probability:.6f}")
```
