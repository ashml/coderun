```python
import sys


def main():
    n, k = map(int, input().split())
    events = []
    for _ in range(n):
        idx, res, team = input().split()
        events.append((int(res), True, team == 'SKI'))
    for _ in range(k):
        idx, res, team = input().split()
        events.append((int(res), False, team == 'SKI'))

    events.sort(key=lambda x: x[0])
    net_balance = 0
    weighted_sum = 0
    current_runners_cnt = 0
    existing_cnt = 0
    runner_counts = []
    for _, is_old, is_ski in events:
        if is_old:
            sign = 1 if is_ski else -1
            weighted_sum += sign * existing_cnt
            net_balance += sign
            if not is_ski:
                current_runners_cnt += 1
            existing_cnt += 1
        else:
            runner_counts.append(current_runners_cnt)

    initial_score = n * net_balance - weighted_sum
    max_score = initial_score
    current_r_sum = 0
    for t, r_val in enumerate(runner_counts, 1):
        current_r_sum += r_val
        candidate = initial_score + n * t + (t * (t + 1)) // 2 - 2 * current_r_sum
        if candidate > max_score:
            max_score = candidate

    print(max_score)


if __name__ == "__main__":
    main()
```
