```python
import random


def simulate(n_problems, n_simulations=1000000):
    n_coins = 2 * n_problems
    letters = ['А', 'Н', 'Л', 'И', 'Т', 'К']
    required_letters = {'А': 3, 'Н': 1, 'Л': 1, 'И': 2, 'Т': 1, 'К': 1}
    success_count = 0

    for _ in range(n_simulations):
        collected_letters = random.choices(letters, k=n_coins)
        letter_counts = {letter: collected_letters.count(letter) for letter in letters}
        if all(letter_counts[letter] >= required_letters[letter] for letter in required_letters):
            success_count += 1

    probability = success_count / n_simulations
    return probability

for n_problems in range(10, 12):
    probability = simulate(n_problems)
    print(f"Вероятность составить слово 'АНАЛИТИКА' при решении {n_problems} задач: {probability:.5f}")
print(f"Ответ: {probability + n_problems:.3f}")
```
