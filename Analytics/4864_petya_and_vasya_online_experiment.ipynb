```python
import sys


def calculate_probabilities(n, p):
    shift = n
    size = 2 * n + 1
    dp_prev = [[0.0] * size for _ in range(2)]
    dp_prev[0][shift + 0] = 1.0 - p
    dp_prev[1][shift + 0] = p

    for _ in range(2, n + 1):
        dp_next = [[0.0] * size for _ in range(2)]
        for prev in (0, 1):
            for d in range(size):
                prob = dp_prev[prev][d]
                if prob == 0.0:
                    continue
                prob0 = prob * (1.0 - p)
                prob1 = prob * p
                if prev == 1:
                    dp_next[0][d + 1] += prob0
                    dp_next[1][d - 1] += prob1
                else:
                    dp_next[0][d] += prob0
                    dp_next[1][d] += prob1
        dp_prev = dp_next

    petya = 0.0
    draw = 0.0
    vasya = 0.0
    for prev in (0, 1):
        for di in range(size):
            prob = dp_prev[prev][di]
            if di > shift:
                petya += prob
            elif di == shift:
                draw += prob
            else:
                vasya += prob
    return petya, draw, vasya


def main():
    n = int(input())
    p = float(input())
    petya, draw, vasya = calculate_probabilities(n, p)
    print("{:.12f} {:.12f} {:.12f}".format(petya, draw, vasya))


if __name__ == '__main__':
    main()
```
