```python
import sys


def main():
    n, k = map(int, input().split())
    tasks = []
    for i in range(n):
        data = input().split()
        tasks.append({'t': int(data[0]), 'd': int(data[1]), 'id': i + 1})
        
    tasks.sort(key=lambda x: x['d'])
    worker_times = [0] * k
    worker_assignments = [[] for _ in range(k)]

    def backtrack(task_idx):
        if task_idx == n:
            return True

        current_task = tasks[task_idx]
        duration = current_task['t']
        deadline = current_task['d']
        original_id = current_task['id']
        seen_loads = set()

        for i in range(k):
            current_load = worker_times[i]

            if current_load in seen_loads:
                continue
            
            if current_load + duration <= deadline:
                worker_times[i] += duration
                worker_assignments[i].append(original_id)
                seen_loads.add(current_load)
                if backtrack(task_idx + 1):
                    return True

                worker_assignments[i].pop()
                worker_times[i] -= duration
            else:
                seen_loads.add(current_load)
        
        return False

    if backtrack(0):
        print('YES')
        for i in range(k):
            output_line = [len(worker_assignments[i])] + worker_assignments[i]
            print(*(output_line))
    else:
        print('NO')


if __name__ == '__main__':
    main()
```
