```python
import random


def simulate_game():
    rolls = [random.randint(1, 6) for _ in range(125)]
    total_prize = 0
    i = 0
    while i < len(rolls):
        length = 1
        while i + length < len(rolls) and rolls[i + length] == rolls[i + length - 1] + 1:
            length += 1
        if length >= 3:
            total_prize += 100 * (length - 2)
        i += length if length > 1 else 1
    return total_prize


def monte_carlo_simulation(num_simulations):
    total_prize = 0
    for _ in range(num_simulations):
        total_prize += simulate_game()
    average_prize = total_prize / num_simulations
    return average_prize


num_simulations = 300000
average_prize = monte_carlo_simulation(num_simulations)
max_cost = int(average_prize)


print(f"Средний выигрыш за игру: {average_prize:.0f} рублей")
print(f"Максимальная целочисленная цена участия: {max_cost - 1} рублей")
```
