```python
import sys


def best_trip(temps):
    min_temp = temps[0]
    min_day = 0
    max_increase = 0
    best_start = 0
    best_end = 0
    
    for i in range(1, len(temps)):
        if temps[i] - min_temp > max_increase:
            max_increase = temps[i] - min_temp
            best_start = min_day
            best_end = i
        if temps[i] < min_temp:
            min_temp = temps[i]
            min_day = i
    return max_increase, best_start, best_end
    

def main():
    temperatures = list(map(int, input().split()))
    result = best_trip(temperatures)
    print(*result)


if __name__ == '__main__':
    main()
```
