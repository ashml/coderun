```python
import sys


def get_digit_mask(num):
    mask = 0
    while num > 0:
        digit = num % 10
        mask |= (1 << digit)
        num //= 10
    return mask


def find_triplet(nums, masks):
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                combined_mask = masks[i] | masks[j] | masks[k]
                if combined_mask == 1023:
                    return f"{nums[i]} {nums[j]} {nums[k]}"
    return None


def main():
    t = int(input())
    results = []
    for _ in range(t):
        n = int(input())
        nums = list(map(int, input().split()))
        masks = [get_digit_mask(num) for num in nums]
        result = find_triplet(nums, masks)
        results.append(result)
    print('\n'.join(results))


if __name__ == '__main__':
    main()
```
