```python
import pandas as pd


file_path = 'data/181_formation_of_swims/'
file_name = 'heats.csv'
df = pd.read_csv(file_path + file_name)

latecomers = df[df['heat'] == 0].copy()
assigned_swimmers = df[df['heat'] != 0].copy()

all_swimmers = pd.concat([latecomers, assigned_swimmers], ignore_index=True)
all_swimmers_sorted = all_swimmers.sort_values(by='time', ascending=False).reset_index(drop=True)

num_swimmers = len(all_swimmers_sorted)
heats = []
heat_num = 1

for i in range(0, num_swimmers, 8):
    heat_swimmers = all_swimmers_sorted.iloc[i:i+8].copy()
    heat_swimmers['heat'] = heat_num
    heats.append(heat_swimmers)
    heat_num += 1

lane_order = [4, 5, 3, 6, 2, 7, 1, 8]

for idx, heat_swimmers in enumerate(heats):
    num_swimmers_in_heat = len(heat_swimmers)
    lanes = lane_order[:num_swimmers_in_heat]
    heat_swimmers_sorted = heat_swimmers.sort_values(by='time').reset_index(drop=True)
    heat_swimmers_sorted['lane'] = lanes
    heats[idx] = heat_swimmers_sorted

all_swimmers_assigned = pd.concat(heats, ignore_index=True)
latecomers_assigned = all_swimmers_assigned[all_swimmers_assigned['id'].isin(latecomers['id'])].copy()
latecomers_assigned = latecomers_assigned.sort_values(by='id')

for _, row in latecomers_assigned.iterrows():
    print(f"{int(row['id'])},{int(row['heat'])},{int(row['lane'])}")
```
