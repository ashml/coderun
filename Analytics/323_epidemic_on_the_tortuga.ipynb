```python
import numpy as np


total_people = 124
initial_infected = 1
binomial_n = 10
binomial_p = 0.2
isolation_day = 14
simulation_count = 10000


def simulate_outbreak():
    days = 0
    infected = [isolation_day] * initial_infected
    visited_hold = set(range(initial_infected))

    while len(visited_hold) < total_people:
        new_infected = 0
        for i in range(len(infected)):
            if infected[i] > 0:
                new_infected += np.random.binomial(binomial_n, binomial_p)
            infected[i] -= 1
        
        for _ in range(new_infected):
            if len(visited_hold) < total_people:
                visited_hold.add(len(visited_hold))
                infected.append(isolation_day)

        days += 1
    return days


results = [simulate_outbreak() for _ in range(simulation_count)]
unique, counts = np.unique(results, return_counts=True)
most_likely_day = unique[np.argmax(counts)]
print(most_likely_day + isolation_day)
```
