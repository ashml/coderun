```python
import sys
from collections import defaultdict


def calculate_probabilities(n):
    memo = {}
    def dp(n, prev, count_OO, count_OR):
        if n == 0:
            return (count_OR > count_OO, count_OR == count_OO, count_OR < count_OO)
        
        key = (n, prev, count_OO, count_OR)
        if key in memo:
            return memo[key]

        if prev == 'O':
            p1, t1, v1 = dp(n - 1, 'O', count_OO + 1, count_OR)
            p2, t2, v2 = dp(n - 1, 'R', count_OO, count_OR + 1)
        else:
            p1, t1, v1 = dp(n - 1, 'O', count_OO, count_OR)
            p2, t2, v2 = dp(n - 1, 'R', count_OO, count_OR)

        memo[key] = (p1 + p2, t1 + t2, v1 + v2)
        return memo[key]

    p_O, t_O, v_O = dp(n - 1, 'O', 0, 0)
    p_R, t_R, v_R = dp(n - 1, 'R', 0, 0)

    total_sequences = 2**n
    petya_win_prob = (p_O + p_R) / total_sequences
    draw_prob = (t_O + t_R) / total_sequences
    vasya_win_prob = (v_O + v_R) / total_sequences

    return petya_win_prob, draw_prob, vasya_win_prob


def main():
    n = int(input())
    petya_win, draw, vasya_win = calculate_probabilities(n)
    print(f"{petya_win:.9f} {draw:.9f} {vasya_win:.9f}")


if __name__ == '__main__':
    main()
```
