```python
import sys


def main():
    n = int(input())
    k = int(input())
    a = [int(input()) for _ in range(n)]
    vals = sorted(set(a))
    comp = {v: i + 1 for i, v in enumerate(vals)}
    m = len(vals)
    bit = [0] * (m + 1)

    def bit_add(i):
        while i <= m:
            bit[i] += 1
            i += i & -i

    def bit_sum(i):
        s = 0
        while i > 0:
            s += bit[i]
            i -= i & -i
        return s

    ans = -1
    for i in range(n):
        if i >= k:
            cnt_less = bit_sum(comp[a[i]] - 1)
            if cnt_less >= (9 * i + 9) // 10:
                ans = i
                break
        bit_add(comp[a[i]])

    print(ans)


if __name__ == '__main__':
    main()
```
