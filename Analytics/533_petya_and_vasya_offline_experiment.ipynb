```python
n = 1000
nmax = n
dp = [[[0.0 for _ in range(2 * nmax + 1)] for _ in range(3)] for _ in range(2)]

symbol_to_index = {'N': 0, 'O': 1, 'R': 2}
index_to_symbol = ['N', 'O', 'R']

dp[0][symbol_to_index['N']][n] = 1.0

for pos in range(1, n + 1):
    curr = pos % 2
    prev = (pos - 1) % 2
    for i in range(3):
        for j in range(2 * nmax + 1):
            dp[curr][i][j] = 0.0

    for prev_index in range(3):
        for D_index in range(0, 2 * nmax + 1):
            prob = dp[prev][prev_index][D_index]
            if prob == 0.0:
                continue
            prev_symbol = index_to_symbol[prev_index]
            D = D_index - nmax

            for c in ['O', 'R']:
                if prev_symbol == 'O' and c == 'O':
                    D_new = D - 1
                elif prev_symbol == 'O' and c == 'R':
                    D_new = D + 1
                else:
                    D_new = D
                if D_new < -nmax or D_new > nmax:
                    continue
                D_new_index = D_new + nmax
                curr_prev_index = symbol_to_index[c]
                dp[curr][curr_prev_index][D_new_index] += prob * 0.5


P = [0.0 for _ in range(-nmax, nmax + 1)]
final = n % 2
for D_index in range(0, 2 * nmax + 1):
    D = D_index - nmax
    P[D] = dp[final][symbol_to_index['O']][D_index] + dp[final][symbol_to_index['R']][D_index]

P_Petya = sum(P[D] for D in range(1, nmax + 1))
P_Tie = P[0]
P_Vasya = sum(P[D] for D in range(-nmax, 0))
print(f"{P_Petya:.10f} {P_Tie:.10f} {P_Vasya:.10f}")
```
