```python
import pandas as pd


file_path = 'data/333_sampled_log/sampled-logs.tsv'
data = pd.read_csv(file_path, sep='\t')

user_day_counts = data.groupby('UUID')['fielddate'].nunique()

one_day_users = (user_day_counts == 1).sum()
two_days_users = (user_day_counts == 2).sum()
three_days_users = (user_day_counts == 3).sum()

print(f"Пользователей, попавших в выборку 1 раз: {one_day_users}")
print(f"Пользователей, попавших в выборку 2 раза: {two_days_users}")
print(f"Пользователей, попавших в выборку 3 раза: {three_days_users}")
```
