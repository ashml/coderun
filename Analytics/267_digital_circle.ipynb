```python
def max_sum_products(max_fives):
    max_fives = min(max_fives, 2021)
    sequence = []

    while max_fives > 1:
        sequence.extend([5, 5, 2])
        max_fives -= 2
    if max_fives == 1:
        sequence.append(5)

    sequence.extend([2] * (len(sequence) % 3))
    n = len(sequence)
    total_sum = 0
    for i in range(n):
        total_sum += sequence[i] * sequence[(i + 1) % n] * sequence[(i + 2) % n]
    return total_sum


result = max_sum_products(2021)
print(result)
```
