```python
import sys
import numpy as np


def main():
    n = int(input())
    transitions = np.array([[int(num) for num in input().split()] for _ in range(n)], dtype='float')
    row_sums = transitions.sum(axis=1)
    P = transitions / row_sums[:, None]
    A = P.T - np.eye(n)
    A = np.vstack([A, np.ones(n)])
    b = np.zeros(n + 1)
    b[-1] = 1
    probs, residuals, rank, s = np.linalg.lstsq(A, b, rcond=None)
    print(' '.join(f"{p:.2f}" for p in probs))


if __name__ == '__main__':
    main()
```
