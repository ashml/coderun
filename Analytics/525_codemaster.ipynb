```python
import sys


def find_index(n, a):
    max_ai = 100000
    freq = [0] * (max_ai + 1)
    for ai in a:
        freq[ai] += 1
    
    sorted_ai = []
    for ai in range(max_ai, -1, -1):
        sorted_ai.extend([ai] * freq[ai])
    
    max_k = min(n, int(max_ai**0.5))
    
    for k in range(1, max_k + 1):
        if sorted_ai[k - 1] < k * k:
            return k - 1
    return max_k


def main():
    n = int(input())
    a = list(map(int, input().split()))
    print(find_index(n, a))


if __name__ == '__main__':
    main()
```
