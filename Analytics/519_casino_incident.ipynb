```python
import sys


def eratosfen(n):
    if n < 2:
        return 0
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False
    p = 2
    while p * p <= n:
        if is_prime[p]:
            for i in range(p * p, n + 1, p):
                is_prime[i] = False
        p += 1
    return sum(is_prime)


def main():
    n = int(input())
    print(f'{eratosfen(n) / n:.2f}')
    

if __name__ == '__main__':
    main()
```
