```python
import sys


def main():
    n, k = map(int, input().split())
    a = [int(el) for el in input().split()]
    unique = sorted(set(a))
    n_unique = len(unique)

    if k >= n_unique:
        result = unique
        curr = unique[-1] + 1
        while len(result) < k:
            result.append(curr)
            curr += 1
        print(*(result))
        return

    cnt = []
    for val in unique:
        c = 0
        for el in a:
            if el >= val:
                c += 1
        cnt.append(c)

    dp = [[-1] * n_unique for _ in range(k + 1)]
    parent = [[-1] * n_unique for _ in range(k + 1)]
    
    for i in range(n_unique):
        dp[1][i] = unique[i] * cnt[i]
        
    for k in range(2, k + 1):
        for i in range(k - 1, n_unique):
            for j in range(k - 2, i):
                gain = (unique[i] - unique[j]) * cnt[i]
                current_score = dp[k-1][j] + gain
                
                if current_score > dp[k][i]:
                    dp[k][i] = current_score
                    parent[k][i] = j

    max_score = -1
    last_idx = -1
    for i in range(k - 1, n_unique):
        if dp[k][i] > max_score:
            max_score = dp[k][i]
            last_idx = i

    path = []
    curr_k = k
    curr_idx = last_idx
    
    while curr_k > 0:
        path.append(unique[curr_idx])
        curr_idx = parent[curr_k][curr_idx]
        curr_k -= 1

    print(*(path[::-1]))


if __name__ == '__main__':
    main()
```
