```python
import sys


def can_get_mass(n, weights):
    achievable_weights = {0}
    
    for weight in weights:
        new_weights = set()
        for aw in achievable_weights:
            new_weights.add(aw + weight)
            new_weights.add(abs(aw - weight))
        achievable_weights.update(new_weights)

    for i in range(1, n + 1):
        if i not in achievable_weights:
            return 'No'
    return 'Yes'


def main():
    n = int(input())
    weights = list(map(int, input().split()))
    print(can_get_mass(n, weights))


if __name__ == '__main__':
    main()
```
