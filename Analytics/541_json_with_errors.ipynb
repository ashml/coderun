```python
import json


def find_incorrect_scores(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)

    incorrect_users = []
    for user_data in data:
        calculated_score = sum(value for key, value in user_data.items() if key != 'username' and key != 'score')
        if calculated_score != user_data['score']:
            incorrect_users.append(user_data['username'])
    return sorted(incorrect_users)


file_path = 'data/541_json_with_errors/test_data.json'
incorrect_users = find_incorrect_scores(file_path)
print('\n'.join(incorrect_users))
```
