```python
import sys


def find_winner(n, m, a):
    prefix = [0] * (n + 1)
    dp = [0] * (n + 1)
    for i in range(n):
        prefix[i+1] = prefix[i] + a[i]
    
    for i in range(n - 1, -1, -1):
        dp[i] = float('-inf')
        for k in range(1, m + 1):
            if i + k > n:
                break
            current = prefix[i + k] - prefix[i]
            dp[i] = max(dp[i], current - dp[i + k])
    return int(dp[0] > 0)


def main():
    n, m = map(int, input().split())
    a = list(map(int, input().split()))
    print(find_winner(n, m, a))


if __name__ == '__main__':
    main()
```
