```python
import sys
from math import comb


def calculate_expected_overdue(n, k, t, p):
    expected_overdue = 0
    for i in range(n + 1):
        prob_k_tomorrow = comb(n, i) * ((1 - p) ** i) * (p ** (n - i))
        if i + k > n:
            overdue = i + k - n
            expected_overdue += overdue * prob_k_tomorrow * t
    return expected_overdue


def main():
    n, k, t = [int(el) for el in input().split()]
    p = float(input())
    expected_overdue = calculate_expected_overdue(n, k, t, p)
    print(f"{expected_overdue:.6f}")


if __name__ == '__main__':
    main()
```
