```python
import sys
sys.setrecursionlimit(10**6)
from collections import defaultdict


def find_paths(matrix, n):
    paths_dict = defaultdict(int)

    def dfs(i, j, path):
        if i < 0 or i >= n or j < 0 or j >= n:
            return
            
        if path and matrix[i][j] <= path[-1]:
            return
        
        new_path = path + [matrix[i][j]]
        if i == 0 or j == 0 or i == n - 1 or j == n - 1:
            paths_dict[len(new_path)] += 1
    
        dfs(i + 1, j, new_path)
        dfs(i - 1, j, new_path)
        dfs(i, j + 1, new_path)
        dfs(i, j - 1, new_path)
    
    for i in range(n):
        for j in range(n):
            dfs(i, j, [])
            
    return paths_dict 


def main():
    n = int(input())
    matrix = [[int(el) for el in input().split()] for _ in range(n)]
    paths_dict = find_paths(matrix, n)
    max_length = max(paths_dict.keys())
    max_count = paths_dict[max_length]
    print(max_length, max_count)
    

if __name__ == '__main__':
    main()
```
