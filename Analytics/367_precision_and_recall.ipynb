```python
import sys
import re


def main():
    pattern = re.compile(r'^([2-9])-(\d{4})(BIG|BAG|BUG|BG|UG)$')
    def is_valid(s):
        match = pattern.match(s)
        if not match:
            return False
        
        region = match.group(1)
        number = match.group(2)
        letters = match.group(3)

        if number == '0000':
            return False

        if region == '5' and letters in ['BIG', 'BAG']:
            return False
            
        return True

    tp = 0
    fp = 0
    n, k = map(int, input().split())
    for i in range(1, n + 1):
        if is_valid(input()):
            tp += 1

    for i in range(k):
        if is_valid(input()):
            fp += 1

    recall = tp / n
    precision = tp / (tp + fp)
    print(f"{recall:.4f} {precision:.4f}")


if __name__ == '__main__':
    main()
```
