```python
import sys


def kadane(arr):
    max_sum = float('-inf')
    current_sum = 0
    for value in arr:
        current_sum += value
        if current_sum > max_sum:
            max_sum = current_sum
        if current_sum < 0:
            current_sum = 0
    return max_sum


def main():
    n, m = [int(el) for el in input().split()]
    matrix = [[int(el) for el in input().split()] for _ in range(n)] 
    max_profit = float('-inf')
    for top in range(n):
        temp = [0] * m
        for bottom in range(top, n):
            for col in range(m):
                temp[col] += matrix[bottom][col]
            current_max = kadane(temp)
            if current_max > max_profit:
                max_profit = current_max
    print(max_profit)


if __name__ == '__main__':
    main()
```
