```python
import sys


def main():
    n = int(input())
    intervals = [tuple(map(int, input().split())) for _ in range(n)]
    intervals.sort(key=lambda x: x[0])
    
    stack = []
    root_color = -1
    possible = True
    
    for l, r, c in intervals:
        if not possible:
            break

        while stack and stack[-1][0] < r:
            top_r, top_c = stack[-1]
            if top_c != c:
                possible = False
                break
            stack.pop()

        if not stack:
            if root_color != -1 and root_color != c:
                possible = False
            else:
                root_color = c

        stack.append((r, c))
    
    print('YES' if possible else 'NO')


if __name__ == "__main__":
    main()
```
