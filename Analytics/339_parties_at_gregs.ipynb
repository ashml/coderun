```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return x / (x**2 - 5*x + 10)

x_values = np.linspace(0, 10, 1000)
y_values = f(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label=r"$\frac{x}{x^2 - 5x + 10}$")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("График функции $\\frac{x}{x^2 - 5x + 10}$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
p = f(4)
p_bombastic = 1 - (1 - p)**4
prob_all_bombastic = p_bombastic**52
print(f"{prob_all_bombastic:.3f}")
```
