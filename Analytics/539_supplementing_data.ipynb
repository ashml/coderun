```python
import pandas as pd
import numpy as np
import os


train_path = 'data/539_supplementing_data/train_00_00.csv'
test_path = 'data/539_supplementing_data/test_00_00.csv'
output_path = 'data/539_supplementing_data/test_filled_result.csv'

train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

test_data.fillna(0, inplace=True)

if 'score' in test_data.columns:
    test_data.drop(columns=['score'], inplace=True)

days = sorted(set(col.split('_')[0] for col in test_data.columns if '_' in col))

for day in days:
    day_columns = [col for col in test_data.columns if col.startswith(day + '_')]
    test_data[f'score_{day}'] = test_data[day_columns].max(axis=1)

score_columns = [f'score_{day}' for day in days]
test_data['score'] = test_data[score_columns].sum(axis=1)
test_data.sort_values(by=['score', 'username'], ascending=[False, True], inplace=True)
test_data.to_csv(output_path, index=False, float_format='%.0f')
```
