```python
from math import comb


def min_players(n=19, target_prob=0.95) -> int:
    if target_prob <= 0.0:
        return 1
    if target_prob >= 1.0:
        return n + 1

    total = 2**n
    cum = 0
    for k in range(1, n + 2):
        cum += comb(n, k - 1)
        prob_success = cum / total
        if prob_success > target_prob:
            return k
    return n + 1


n = 19
p = 0.95
print(f"При N={n} и p={p:.2f} минимальное количество игроков -> k={min_players(n, p)}")
```
