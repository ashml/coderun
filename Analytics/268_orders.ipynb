```python
from fractions import Fraction

def calculate_expected_weeks():
    restaurant_combinations = 21
    store_orders = 7
    total_combinations = restaurant_combinations * store_orders
    favorable_combinations = 3 * 3
    probability = Fraction(favorable_combinations, total_combinations)
    expected_weeks = Fraction(1, probability)
    result = expected_weeks.numerator + expected_weeks.denominator
    return result

print(calculate_expected_weeks())
```
