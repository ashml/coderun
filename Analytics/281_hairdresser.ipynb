```python
import sys


def min_expected_loss(n, k, probabilities):
    total_loss = sum(probabilities)
    if k == 0:
        return total_loss
    if k == n:
        return 0.0
    
    current_window_loss = sum(probabilities[:k])
    min_loss = total_loss - current_window_loss
    
    for i in range(k, n):
        current_window_loss += probabilities[i] - probabilities[i - k]
        min_loss = min(min_loss, total_loss - current_window_loss)
    
    return min_loss
    

def main():
    n = int(input())
    k = int(input())
    probabilities = [float(input()) for _ in range(n)]
    print(min_expected_loss(n, k, probabilities))


if __name__ == '__main__':
    main()
```
