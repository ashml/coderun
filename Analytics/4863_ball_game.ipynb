```python
import sys
from fractions import Fraction


def probability_game_ends(x, y, n):
    fraction_result = Fraction(1, 1)
    for i in range(1, n + 1):
        curr = i * (i + x + y)
        fraction_result *= Fraction(curr, curr + x * y)
    return fraction_result.denominator - fraction_result.numerator, fraction_result.denominator


def main():
    n = int(input())
    x = int(input())
    y = int(input())
    p, q = probability_game_ends(x, y, n)
    print(f"{p} {q}")


if __name__ == '__main__':
    main()
```
