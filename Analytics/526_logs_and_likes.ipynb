```python
import pandas as pd
import ast

file_path = "data/526_logs_and_likes/logs.tsv"
logs = pd.read_csv(file_path, sep='\t', low_memory=False)
like_logs = logs[(logs['event'] == 'like') & (logs.context != '[]')]
```

```python
def parse_context(s):
    return ast.literal_eval(s)

def extract_art_id(s):
    return s[s.rfind('/') + 1:-2]

like_logs["paths"] = like_logs["context"].apply(parse_context)
df_exploded = like_logs.explode("paths").rename(columns={"paths": "path"})
df_result = df_exploded[["user", "path"]].reset_index(drop=True)
df_result['art_id'] = df_result['path'].apply(extract_art_id)
result = df_result[['user', 'art_id']].drop_duplicates()
```

```python
counts = result["art_id"].value_counts()
top3 = counts.head(3)

print(top3)
```
