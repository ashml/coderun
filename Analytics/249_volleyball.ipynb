```python
def compute_set_ways(a, b, target):
    dp = [[[0 for _ in range(2)] for _ in range(b + 2)] for _ in range(a + 2)]
    dp[0][0][1] = 1

    for i in range(a + 1):
        for j in range(b + 1):
            for ongoing in [0, 1]:
                ways = dp[i][j][ongoing]
                if ways == 0:
                    continue

                finished = False
                if ongoing == 1:
                    if (i >= target or j >= target) and abs(i - j) >= 2:
                        dp[i][j][0] += ways
                        finished = True

                if ongoing == 0 or finished:
                    continue

                if i + 1 <= a:
                    if (i + 1 >= target or j >= target) and abs((i + 1) - j) >= 2:
                        dp[i + 1][j][0] += ways
                    else:
                        dp[i + 1][j][1] += ways

                if j + 1 <= b:
                    if (i >= target or j + 1 >= target) and abs(i - (j + 1)) >= 2:
                        dp[i][j + 1][0] += ways
                    else:
                        dp[i][j + 1][1] += ways
    return dp[a][b][0]


def main():
    data = input().split()
    n = int(data[0])
    scores = data[1:]

    total_ways = 1
    target = 25
    for idx, score in enumerate(scores):
        score_a, score_b = map(int, score.split(':'))
        if idx == 4:
            target = 15
        ways = compute_set_ways(score_a, score_b, target)
        total_ways *= ways
    print(total_ways)


if __name__ == "__main__":
    main()
```
