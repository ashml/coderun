```python
import sys


def main():
    t = int(input())
    results = []
    for _ in range(t):
        k, x, y, v, u, a, b = map(int, input().split())
    
        total = 37
        p_p = (y - x + 1) / total
        p_v = (u - v + 1) / total
        
        p_both = max(0, min(y, u) - max(x, v) + 1) / total
        p_only_p = p_p - p_both
        p_only_v = p_v - p_both
        p_none = 1 - (p_both + p_only_p + p_only_v)

        dp = [[[0.0] * (b + 2) for i in range(a + 2)] for j in range(k + 1)]
        dp[0][0][0] = 1.0
        
        for i in range(k):
            for j in range(a + 1):
                for z in range(b + 1):
                    cur = dp[i][j][z]
                    if cur == 0: 
                        continue
                    dp[i+1][j][z] += cur * p_none
                    if j+1 <= a:
                        dp[i+1][j+1][z] += cur * p_only_p
                    if z+1 <= b:
                        dp[i+1][j][z+1] += cur * p_only_v
                    if j+1 <= a and z+1 <= b:
                        dp[i+1][j+1][z+1] += cur * p_both
        
        results.append(dp[k][a][b])
    
    print('\n'.join(map(str, results)))


if __name__ == '__main__':
    main()
```
