```python
import sys


def main():
    INF = float('inf')
    NINF = float('-inf')

    corners_ne = []
    corners_nw = []
    corners_se = []
    corners_sw = []
    
    xs = set()
    n = int(input())
    for _ in range(n):
        x, y, v_dir, h_dir = map(int, input().split())
        xs.add(x)
        
        if v_dir == 1 and h_dir == 1:
            corners_ne.append((x, y))
        elif v_dir == 1 and h_dir == -1:
            corners_nw.append((x, y))
        elif v_dir == -1 and h_dir == 1:
            corners_se.append((x, y))
        elif v_dir == -1 and h_dir == -1:
            corners_sw.append((x, y))

    if not (corners_ne and corners_nw and corners_se and corners_sw):
        print("NO")
        return

    sorted_xs = sorted(list(xs))
    m = len(sorted_xs)
    x_map = {x: i for i, x in enumerate(sorted_xs)}

    ne_at = [[] for _ in range(m)]
    nw_at = [[] for _ in range(m)]
    se_at = [[] for _ in range(m)]
    sw_at = [[] for _ in range(m)]

    for x, y in corners_ne: 
        ne_at[x_map[x]].append(y)
    for x, y in corners_nw: 
        nw_at[x_map[x]].append(y)
    for x, y in corners_se: 
        se_at[x_map[x]].append(y)
    for x, y in corners_sw: 
        sw_at[x_map[x]].append(y)

    pref_ne = [INF] * m
    curr = INF
    for i in range(m):
        for y in ne_at[i]:
            if y < curr: curr = y
        pref_ne[i] = curr

    pref_se = [NINF] * m
    curr = NINF
    for i in range(m):
        for y in se_at[i]:
            if y > curr: curr = y
        pref_se[i] = curr

    suff_nw = [INF] * m
    curr = INF
    for i in range(m - 1, -1, -1):
        for y in nw_at[i]:
            if y < curr: curr = y
        suff_nw[i] = curr
        
    suff_sw = [NINF] * m
    curr = NINF
    for i in range(m - 1, -1, -1):
        for y in sw_at[i]:
            if y > curr: curr = y
        suff_sw[i] = curr

    if suff_sw[0] < suff_nw[0]:
        print('NO')
        return

    if pref_se[-1] < pref_ne[-1]:
        print('NO')
        return

    for i in range(m):
        top_y = min(pref_ne[i], suff_nw[i])
        bot_y = max(pref_se[i], suff_sw[i])
        if bot_y < top_y:
            print('NO')
            return

        if i < m - 1:
            top_y_int = min(pref_ne[i], suff_nw[i+1])
            bot_y_int = max(pref_se[i], suff_sw[i+1])
            if bot_y_int < top_y_int:
                print('NO')
                return

    print('YES')


if __name__ == '__main__':
    main()
```
