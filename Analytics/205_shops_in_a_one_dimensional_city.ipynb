```python
import numpy as np
from scipy.stats import norm

params = np.array([
    (150, 80),
    (310, 60),
    (450, 70),
    (550, 70),
    (730, 85),
    (900, 90)
])

def ca_density(x):
    x = np.atleast_1d(x)[:, None]
    mu, sigma = params.T
    densities = norm.pdf(x, mu, sigma)
    return np.max(densities, axis=1)

def integrate(a, b, dx=1):
    x = np.arange(a, b + dx, dx)
    densities = ca_density(x)
    return np.sum(densities) * dx

def shop_coverage(x):
    return integrate(x - 100, x + 100)


min_x = 100
max_x = 900
x_values = np.arange(min_x, max_x + 1)
coverage = np.array([shop_coverage(x) for x in x_values])

max_total = 0
best_pair = (0, 0)

for i, x1 in enumerate(x_values):
    for j, x2 in enumerate(x_values):
        if x2 >= x1 + 200:
            current_total = coverage[i] + coverage[j]
            if current_total > max_total:
                max_total = current_total
                best_pair = (x1, x2)

print(best_pair[0], best_pair[1])
```
