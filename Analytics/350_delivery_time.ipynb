```python
import numpy as np


num_simulations = 100000000

A = np.random.uniform(10, 30, num_simulations)
B = np.random.uniform(2, 7, num_simulations)
C = np.random.uniform(5, 20, num_simulations)
D = np.random.uniform(5, 15, num_simulations)


E = np.maximum(A, B + C)
T = E + D
T_95 = np.percentile(T, 95)

print(f"95-й процентиль общего времени доставки: {T_95:.3f} минут")
```
