```python
import pandas as pd
import xml.etree.ElementTree as ET


def xml_to_csv(xml_path, output_csv_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    rows = []
    for row in root.findall('Row'):
        data = {}
        for col in row.findall('Column'):
            col_name = col.attrib['name']
            data[col_name] = int(col.text) if col.text.isdigit() else col.text
        rows.append(data)

    df = pd.DataFrame(rows)
    columns_order = ['username', '1_1', '1_2', '1_3', '1_4', 
                     '2_1', '2_2', '2_3', '2_4', 
                     '3_1', '3_2', '3_3', '3_4', 'score']

    df = df.reindex(columns=columns_order).fillna('')
    for col in columns_order[1:]:
        df[col] = df[col].apply(lambda x: int(x) if x != '' else x)
    df = df.sort_values(by=['score', 'username'], ascending=[False, True])
    df.to_csv(output_csv_path, index=False)


xml_to_csv('data/540_xml_back_to_csv/test_data.xml', 'data/540_xml_back_to_csv/test_data_converted.csv')
```
