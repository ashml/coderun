```python
import sys
from itertools import permutations


def degree_of_sequence(a):
    curr = list(a)
    for d in range(len(curr)):
        if all(x == curr[0] for x in curr):
            return d
        curr = [curr[i+1] - curr[i] for i in range(len(curr) - 1)]
    return len(curr) - 1


def main():
    t = int(input())
    out_lines = []
    for _ in range(t):
        n = int(input())
        vals = [int(el) for el in input().split()]
        best = n-1
        for perm in permutations(vals):
            d = degree_of_sequence(perm)
            if d < best:
                best = d
                if best == 0:
                    break
        out_lines.append(str(best))
    print('\n'.join(out_lines))


if __name__ == '__main__':
    main()
```
