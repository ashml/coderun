```python
import sys


def is_prime(num):
    if num < 2: 
        return False
    if num == 2: 
        return True
    if num % 2 == 0: 
        return False
    for i in range(3, int(num**0.5) + 1, 2):
        if num % i == 0:
            return False
    return True

def main():
    n, k = map(int, input().split())
    distance = 1
    while not is_prime(n + distance):
        distance  += 1

    if distance % (k + 1) != 0:
        print('YES')
    else:
        print('NO')


if __name__ == "__main__":
    main()
```
