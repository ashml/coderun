```python
import sys
import json
from collections import defaultdict


def get_dates(data):
    date_max_counts = defaultdict(int)
    for entry in data:
        date = entry['date']
        count = entry['count']
        date_max_counts[date] = max(date_max_counts[date], count)
    
    min_max_count = min(date_max_counts.values())
    result_dates = sorted(date for date, count in date_max_counts.items() if count == min_max_count)
    return result_dates


def main():
    with open('data.json', 'r') as file:
        data = json.load(file)

    if data:
        result_dates = get_dates(data)
        for date in result_dates:
            print(date)


if __name__ == '__main__':
    main()
```
