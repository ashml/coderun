```python
import numpy as np


def check(points, r):
    n = len(points)
    points_sorted = sorted(points, key=lambda p: p[0])
    for i in range(n):
        x_start = points_sorted[i][0]
        x_end = x_start + 2 * r
        temp_points = []
        for j in range(i, n):
            if points_sorted[j][0] <= x_end:
                temp_points.append(points_sorted[j][1])
            else:
                break
        temp_points.sort()
        left = 0
        for right in range(len(temp_points)):
            while temp_points[right] - temp_points[left] > 2 * r:
                left += 1
            if right - left + 1 >= 6:
                return True
    return False


def find_min_radius(points):
    left = 0
    right = max(max(p[0] for p in points) - min(p[0] for p in points),
             max(p[1] for p in points) - min(p[1] for p in points))
    for _ in range(50):
        mid = (left + right) / 2
        if check(points, mid):
            right = mid
        else:
            left = mid
    return right


file_path = 'data/challenge_6_of_36/data.txt'
with open(file_path, 'r') as file:
    data = file.readlines()

results = []
for line in data:
    numbers = list(map(float, line.strip().split()))
    points = list(zip(numbers[::2], numbers[1::2]))
    min_radius = find_min_radius(points)
    results.append(f"{min_radius:.12f}")

for val in results:
    print(val)
```
